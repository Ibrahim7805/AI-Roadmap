```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# 1. Read image with OpenCV (loads as BGR)
image_bgr = cv2.imread('ibrahim.png')

plt.imshow(image_bgr)
plt.axis('off')
plt.show()
```

```python
# 2. Convert from BGR to RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# convert to grayscale
gray_img = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img,cmap='gray')
plt.axis('off')
plt.show()
```

```python
# using HSV (Hue Saturation Value ) color space
hsv_img = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_img)
plt.axis('off')
plt.show()
 
```

```python
# check the shape of the image
print("Image shape:", image_rgb.shape)
```

```python
# resize the image
resized_img = cv2.resize(image_rgb, (400, 400))
plt.imshow(resized_img)
#plt.axis('off')
plt.show()
```

```python
# crop the image
cropped_img = resized_img[50:200, 100:300] #(400 , 400)
plt.imshow(cropped_img)
plt.axis('off')
plt.show()
```

```python
# Canny Edge Detection
edges = cv2.Canny(image_rgb, 100, 200)
# 100 and 200 are the min and max threshold values
plt.imshow(edges)
plt.axis('off')
plt.show()
```

```python
# Canny Edge Detection
edges = cv2.Canny(image_rgb, 100, 200)
# 100 and 200 are the min and max threshold values
plt.imshow(edges , cmap='gray')
plt.axis('off')
plt.show()
```

```python
# gaussian blur
blurred_img = cv2.GaussianBlur(image_rgb, (15 , 15) , 0)
# (15, 15) is the kernel size
# 0 is the sigmaX value (standard deviation in X direction)
plt.imshow(blurred_img)
plt.axis('off')
plt.show()
```

```python
# canny on blurred image
blurred_edges = cv2.Canny(blurred_img, 50,150)
plt.imshow(blurred_edges)
plt.axis('off')
plt.show()
```

```python
# Draw a rectangle on the image
image_with_rectangle = image_rgb.copy()
cv2.rectangle(image_with_rectangle, (200,200), (600,600), (255, 0, 0), 3)
# (50, 50) is the top-left corner
# (200, 200) is the bottom-right corner
# (255, 0, 0) is the color (red in RGB)
# 3 is the thickness of the rectangle border
plt.imshow(image_with_rectangle)
#plt.axis('off')
plt.show()
```

```python
# Draw a circle on the image
image_with_circle = image_rgb.copy()
cv2.circle(image_with_circle, (400,400), 200, (0, 255, 0), -1)
# (300, 300) is the center of the circle
# 50 is the radius
# (0, 255, 0) is the color (green in RGB)
plt.imshow(image_with_circle)
#plt.axis('off')
plt.show()
```

```python
# Put text on the image
image_with_text = image_rgb.copy()
cv2.putText(image_with_text, 'Ibrahim', (120, 1200), cv2.FONT_HERSHEY_SIMPLEX, 5, (255, 255, 255), 6)

# 'Ibrahim' is the text to be added
# (200, 1200) is the bottom-left corner of the text string in the image
# cv2.FONT_HERSHEY_SIMPLEX is the font type
# 2 is the font scale (size)
# (0, 0, 255) is the color (blue in RGB)
# 3 is the thickness of the text

plt.imshow(image_with_text)
plt.axis('off')
plt.show()
# Save the modified image
cv2.imwrite('cat_with_text.jpg', cv2.cvtColor(image_with_text, cv2.COLOR_RGB2BGR))

```

```python
# compare between canny and blur image
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title('Canny Edge Detection')
plt.imshow(edges , cmap='gray')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Blurred Image Canny')
plt.imshow(blurred_edges , cmap='gray')
plt.axis('off')
plt.show()
```

```python
capture = cv2.VideoCapture('dog.mp4')
while True:
    red , frame = capture.read()

    if not red: #
        break

    resized_frame = cv2.resize(frame,(400,400))
    cv2.imshow('Dog', resized_frame)

    if cv2.waitKey(10) & 0xFF==ord('q'):
        break

capture.release()
cv2.destroyAllWindows()
```

```python
capture = cv2.VideoCapture('dog.mp4')

while True:
    red , frame = capture.read()

    if not red:
        capture.set(cv2.CAP_PROP_POS_FRAMES,0)
        # CAP_PROP_POS_FRAMES is used to set the position of the next frame to be read
        continue
        

    resized_frame = cv2.resize(frame,(400,400))
    cv2.imshow('Video', resized_frame)

    if cv2.waitKey(10) & 0xFF==ord('q'):
        # Press 'q' to quit
        # 0xFF==ord('q') this condition checks if the 'q' key is pressed
        break

capture.release()
cv2.destroyAllWindows()
```
