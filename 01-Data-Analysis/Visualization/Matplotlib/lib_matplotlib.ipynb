```python
import pandas as pd
import numpy as np

# ÿπÿØÿØ ÿßŸÑÿ¥ŸáŸàÿ± ÿßŸÑŸÑŸä ŸáŸÜÿ¥ÿ™ÿ∫ŸÑ ÿπŸÑŸäŸáÿß
n_months = 50

# ÿ•ŸÜÿ¥ÿßÿ° ÿ™Ÿàÿßÿ±ŸäÿÆ ÿ¥Ÿáÿ±Ÿäÿ© ÿ™ÿ®ÿØÿ£ ŸÖŸÜ ŸäŸÜÿßŸäÿ± 2000
dates = pd.date_range(start="2000-01-01", periods=n_months, freq="MS")

# ÿ™ŸàŸÑŸäÿØ ÿ®ŸäÿßŸÜÿßÿ™ CO2 (ÿ®ÿØÿßŸäÿ© ŸÖŸÜ 370 ÿ®ÿ≤ŸäÿßÿØÿ© ÿ™ÿØÿ±Ÿäÿ¨Ÿäÿ© + ÿ∂Ÿàÿ∂ÿßÿ° ÿπÿ¥Ÿàÿßÿ¶Ÿäÿ© ÿ®ÿ≥Ÿäÿ∑ÿ©)
co2 = 370 + np.linspace(0, 10, n_months) + np.random.normal(0, 0.5, n_months)

# ÿ™ŸàŸÑŸäÿØ ÿ®ŸäÿßŸÜÿßÿ™ ÿØÿ±ÿ¨ÿ© ÿßŸÑÿ≠ÿ±ÿßÿ±ÿ© ÿßŸÑŸÜÿ≥ÿ®Ÿäÿ© (ÿ™ÿ™ÿ∞ÿ®ÿ∞ÿ® ÿπŸÑŸâ ÿ¥ŸÉŸÑ ŸÖŸàÿ¨ÿ© ÿ≥ŸÜŸàŸäÿ©)
relative_temp = 0.1 * np.sin(np.linspace(0, 4 * np.pi, n_months)) + np.random.normal(0, 0.02, n_months)

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿØÿßÿ™ÿß ŸÅÿ±ŸäŸÖ
data = {
    "date": dates,
    "co2": co2,
    "relative_temp": relative_temp
}

df = pd.DataFrame(data)
df.set_index("date", inplace=True)
df
```

## **Time-Series Data**

ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸÄ Time-Series = ÿ®ŸäÿßŸÜÿßÿ™ ÿ®ÿ™ÿ™ŸÇÿßÿ≥ ÿ®ÿßŸÜÿ™ÿ∏ÿßŸÖ ÿπŸÑŸâ ŸÅÿ™ÿ±ÿßÿ™ ÿ≤ŸÖŸÜŸäÿ© (ŸäŸàŸÖÿå ÿ¥Ÿáÿ±ÿå ÿ≥ŸÜÿ©...).

ÿ±ÿ≥ŸÖŸáÿß ÿ®ŸäÿÆŸÑŸäŸÜÿß ŸÜÿ¥ŸàŸÅ ÿßŸÑÿßÿ™ÿ¨ÿßŸáÿßÿ™ - ÿßŸÑÿ™ÿ∫Ÿäÿ±ÿßÿ™ - ÿßŸÑŸÖŸàÿ≥ŸÖŸäÿ© - ÿßŸÑŸÇŸÅÿ≤ÿßÿ™.

```python
import pandas as pd
import numpy as np

# ÿπÿØÿØ ÿßŸÑÿ¥ŸáŸàÿ± ÿßŸÑŸÑŸä ŸáŸÜÿ¥ÿ™ÿ∫ŸÑ ÿπŸÑŸäŸáÿß
n_months = 50

# ÿ•ŸÜÿ¥ÿßÿ° ÿ™Ÿàÿßÿ±ŸäÿÆ ÿ¥Ÿáÿ±Ÿäÿ© ÿ™ÿ®ÿØÿ£ ŸÖŸÜ ŸäŸÜÿßŸäÿ± 2000
dates = pd.date_range(start="2000-01-01", periods=n_months, freq="MS")

# ÿ™ŸàŸÑŸäÿØ ÿ®ŸäÿßŸÜÿßÿ™ CO2 (ÿ®ÿØÿßŸäÿ© ŸÖŸÜ 370 ÿ®ÿ≤ŸäÿßÿØÿ© ÿ™ÿØÿ±Ÿäÿ¨Ÿäÿ© + ÿ∂Ÿàÿ∂ÿßÿ° ÿπÿ¥Ÿàÿßÿ¶Ÿäÿ© ÿ®ÿ≥Ÿäÿ∑ÿ©)
co2 = 370 + np.linspace(0, 10, n_months) + np.random.normal(0, 0.5, n_months)

# ÿ™ŸàŸÑŸäÿØ ÿ®ŸäÿßŸÜÿßÿ™ ÿØÿ±ÿ¨ÿ© ÿßŸÑÿ≠ÿ±ÿßÿ±ÿ© ÿßŸÑŸÜÿ≥ÿ®Ÿäÿ© (ÿ™ÿ™ÿ∞ÿ®ÿ∞ÿ® ÿπŸÑŸâ ÿ¥ŸÉŸÑ ŸÖŸàÿ¨ÿ© ÿ≥ŸÜŸàŸäÿ©)
relative_temp = 0.1 * np.sin(np.linspace(0, 4 * np.pi, n_months)) + np.random.normal(0, 0.02, n_months)

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿØÿßÿ™ÿß ŸÅÿ±ŸäŸÖ
data = {
    "date": dates,
    "co2": co2,
    "relative_temp": relative_temp
}

df = pd.DataFrame(data)
df.set_index("date", inplace=True)
df
```

```python
fig, ax = plt.subplots()

ax.plot(df.index, df["relative_temp"])

ax.set_xlabel("Time")
ax.set_ylabel("Relative temperature (Celsius)")

plt.show()
```

## **Annotating time-series data**

‚úÖ ŸÑŸäŸá ŸÜÿ∂ŸäŸÅ ÿ™ÿπŸÑŸäŸÇÿßÿ™ (annotations) ÿπŸÑŸâ ÿßŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿ®ŸäÿßŸÜŸäÿü

ÿπÿ¥ÿßŸÜ ŸÜŸàÿ¨ŸëŸá ÿßŸÜÿ™ÿ®ÿßŸá ÿßŸÑŸÇÿßÿ±ÿ¶ ŸÑÿ≠ÿßÿ¨ÿ© ŸÖŸáŸÖÿ© ŸÅŸä ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ÿå ÿ≤Ÿä:

ŸÇŸÅÿ≤ÿ© ŸÉÿ®Ÿäÿ±ÿ© ŸÅŸä ŸÇŸäŸÖÿ©.

ÿ®ÿØÿßŸäÿ© ÿßÿ™ÿ¨ÿßŸá ÿ¨ÿØŸäÿØ (Trend).

ŸÜŸÇÿ∑ÿ© ÿßÿ≥ÿ™ÿ´ŸÜÿßÿ¶Ÿäÿ© (Outlier).

ÿ≠ÿØÿ´ ŸÖŸáŸÖ (ÿ≤Ÿä: ÿ£ŸàŸÑ ŸÖÿ±ÿ© ÿ™ÿπÿØŸä ÿßŸÑÿ≠ÿ±ÿßÿ±ÿ© 1 ÿØÿ±ÿ¨ÿ©).

```python
ig, ax1  = plt.subplots()

# ÿßŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿ£ŸàŸÑ ŸÑŸÖÿ≠Ÿàÿ± CO2
ax1.plot(df.index, df['co2'], color='b')
ax1.set_ylabel("CO2 (ppm)", color='blue')
ax1.tick_params(axis='y', colors='blue')

#ÿ®ŸÜŸÑŸàŸëŸÜ ÿßŸÑÿ™ŸÉÿßÿ™ (ticks) ÿ®ÿ™ÿßÿπÿ© ÿßŸÑŸÖÿ≠Ÿàÿ± y ÿ®ŸÑŸàŸÜ ÿ£ÿ≤ÿ±ŸÇÿå ŸÑÿ™ÿ£ŸÉŸäÿØ ÿßŸÑÿßÿ±ÿ™ÿ®ÿßÿ∑ ÿ®ŸÄ CO2.

# ÿ•ŸÜÿ¥ÿßÿ° ŸÖÿ≠Ÿàÿ± ÿ™Ÿàÿ£ŸÖ ŸÑŸÖŸÇŸäÿßÿ≥ ŸÖÿÆÿ™ŸÑŸÅ
ax2 = ax1.twinx()

# ÿßŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿ´ÿßŸÜŸä ŸÑŸÖÿ≠Ÿàÿ± ÿØÿ±ÿ¨ÿ© ÿßŸÑÿ≠ÿ±ÿßÿ±ÿ©
ax2.plot(df.index, df['relative_temp'], color='r')
ax2.set_ylabel("Relative temperature (¬∞C)", color='red')
ax2.tick_params(axis='y', colors='red')

# --- ÿßŸÑÿ•ÿ∂ÿßŸÅÿ© ÿßŸÑÿ¨ÿØŸäÿØÿ©: ÿßŸÑÿ™ÿπŸÑŸäŸÇ annotation ---
ax2.annotate(">1 degree",
             xy=(pd.Timestamp("2003-07-01"), 1),
             xytext=(pd.Timestamp("2002-07-01"), -0.1),
             arrowprops={})

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ÿ•ÿπÿßÿØÿ© ÿ•ŸÜÿ¥ÿßÿ° ÿ®ŸäÿßŸÜÿßÿ™ ŸÖŸäÿØÿßŸÑŸäÿßÿ™ ÿ£ŸàŸÑŸÖÿ®ŸäÿßÿØ 2016
data = {
    "Country": ["United States", "Great Britain", "China", "Russia", "Germany", "Japan", "Egypt"],
    "Gold": [46, 27, 26, 19, 17, 12, 3],
    "Silver": [37, 23, 18, 17, 10, 8, 2],
    "Bronze": [38, 17, 26, 20, 15, 21, 3]
}

df = pd.DataFrame(data)
df.set_index("Country", inplace=True)
df
```

### **üìå ÿ£ŸàŸÑŸãÿß: Error Bars**

üü¶ ÿßŸÑŸáÿØŸÅ:

ÿ•ÿ∂ÿßŸÅÿ© ÿÆÿ∑ ÿπŸÖŸàÿØŸä ÿµÿ∫Ÿäÿ± ŸÅŸàŸÇ ÿßŸÑÿ®ÿßÿ± ÿ£Ÿà ÿßŸÑŸÜŸÇÿ∑ÿ© ŸÅŸä ÿßŸÑŸÄ line plotÿå ÿ®ŸäŸàÿ∂ÿ≠ ŸÖÿØŸâ ÿßŸÜÿ™ÿ¥ÿßÿ± ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ (ŸÖÿ´ŸÑŸãÿß ÿßŸÑÿßŸÜÿ≠ÿ±ÿßŸÅ ÿßŸÑŸÖÿπŸäÿßÿ±Ÿä std).

```python
import matplotlib.pyplot as plt
import pandas as pd

# ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ¨ŸÖÿ®ÿßÿ≤
gymnastics_data = {
    "ID": range(1, 11),
    "Name": [f"Gymnast_{i}" for i in range(1, 11)],
    "Sport": ["Gymnastics"] * 10,
    "Height": [165, 168, 170, 166, 169, 167, 171, 172, 169, 168],
    "Weight": [60, 62, 63, 61, 60, 64, 65, 63, 62, 61],
    "Medal": ["Gold", "Silver", "Bronze", "Gold", "Silver", "Bronze", "Gold", "Silver", "Bronze", "Gold"],
    "Country": ["USA", "CHN", "JPN", "RUS", "BRA", "EGY", "GER", "FRA", "ITA", "CAN"]
}

# ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿ¨ÿØŸäŸÅ
rowing_data = {
    "ID": range(11, 21),
    "Name": [f"Rower_{i}" for i in range(1, 11)],
    "Sport": ["Rowing"] * 10,
    "Height": [190, 192, 189, 193, 195, 191, 188, 194, 196, 190],
    "Weight": [90, 92, 89, 93, 95, 91, 88, 94, 96, 90],
    "Medal": ["Gold", "Silver", "Bronze", "Gold", "Silver", "Bronze", "Gold", "Silver", "Bronze", "Gold"],
    "Country": ["USA", "GBR", "AUS", "CAN", "NZL", "NED", "GER", "EGY", "NOR", "SWE"]
}
# ÿ™ÿ≠ŸàŸäŸÑŸáŸÖ ÿ•ŸÑŸâ DataFrames
df_gymnastics = pd.DataFrame(gymnastics_data)
df_rowing = pd.DataFrame(rowing_data)

# ÿØŸÖÿ¨ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
df = pd.concat([df_gymnastics, df_rowing], ignore_index=True)

# ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ŸÖÿ™Ÿàÿ≥ÿ∑ ÿßŸÑÿ∑ŸàŸÑ ŸàÿßŸÑÿßŸÜÿ≠ÿ±ÿßŸÅ ÿßŸÑŸÖÿπŸäÿßÿ±Ÿä ŸÑŸÉŸÑ ÿ±Ÿäÿßÿ∂ÿ©
summary = df.groupby("Sport")["Height"].agg(["mean", "std"]).reset_index()
summary
```

| ÿßŸÑÿÆÿßÿµŸäÿ©                           | ÿßŸÑÿ∫ÿ±ÿ∂                                |
| --------------------------------- | ------------------------------------ |
| `ignore_index=True` ŸÅŸä `concat()` | ŸäÿπŸäÿØ ÿ™ÿ±ŸÇŸäŸÖ ÿßŸÑÿµŸÅŸàŸÅ ÿ®ÿπÿØ ÿßŸÑÿØŸÖÿ¨          |
| `drop=True` ŸÅŸä `reset_index()`    | Ÿäÿ≠ÿ∞ŸÅ ÿßŸÑŸÅŸáÿ±ÿ≥ ÿßŸÑŸÇÿØŸäŸÖ ÿ®ÿØŸÑ ŸÖÿß ŸäÿÆŸÑŸäŸá ÿπŸÖŸàÿØ |


‚ú≥Ô∏è ÿßŸÑŸáÿØŸÅ:

ŸÖŸÇÿßÿ±ŸÜÿ© ŸÖÿ™Ÿàÿ≥ÿ∑ ÿ∑ŸàŸÑ ÿßŸÑŸÑÿßÿπÿ®ŸäŸÜ ÿ®ŸäŸÜ ÿßŸÑÿ±Ÿäÿßÿ∂ÿ™ŸäŸÜ

ŸÑŸÉŸÜ ÿ®ÿØŸÑ ŸÖÿß ŸÜÿπÿ±ÿ∂ ÿßŸÑŸÖÿ™Ÿàÿ≥ÿ∑ ÿ®ÿ≥ÿå ŸáŸÜÿ∂ŸäŸÅ Error Bar ŸÑŸäŸàÿ∂ÿ≠ ŸÖÿØŸâ ÿßŸÑÿ™ÿ¥ÿ™ÿ™ (ÿßŸÑÿßŸÜÿ≠ÿ±ÿßŸÅ ÿßŸÑŸÖÿπŸäÿßÿ±Ÿä).

ŸàÿßŸÑŸÖÿ≥ÿßŸÖŸäÿ± (error bars) ÿ®ÿ™Ÿàÿ∂ÿ≠ ŸÖÿØŸâ ÿßŸÑÿ™ÿ¥ÿ™ÿ™:

yerr=summary["std"] ‚Üí ÿ®ŸäŸàÿ∂ÿ≠ ÿßŸÑÿßŸÜÿ≠ÿ±ÿßŸÅ ÿßŸÑŸÖÿπŸäÿßÿ±Ÿä.

capsize=10 ‚Üí ÿ≠ÿ¨ŸÖ ÿ±ÿ£ÿ≥ ÿßŸÑÿÆÿ∑ ÿßŸÑÿπŸÖŸàÿØŸä (error bar).

color=["skyblue", "salmon"] ‚Üí ŸÑŸàŸÜ ŸÖÿÆÿ™ŸÑŸÅ ŸÑŸÉŸÑ ÿ±Ÿäÿßÿ∂ÿ©.

```python
fig, ax = plt.subplots()

ax.bar(summary["Sport"], summary["mean"], yerr=summary["std"], capsize=10, color=["skyblue", "salmon"])
```

### **üìå ÿØÿßŸÑÿ© errorbar() ŸÅŸä Matplotlib**

‚úÖ ÿßŸÑÿ∫ÿ±ÿ∂:

**ax.errorbar(x, y, yerr=errors)**

| ÿßŸÑŸÖÿπÿßŸÖŸÑ         | Ÿàÿ∏ŸäŸÅÿ™Ÿá                                                     |
| --------------- | ---------------------------------------------------------- |
| `x`             | ŸÖÿ≠Ÿàÿ± X (ŸÖÿ´ŸÑÿßŸã ÿßŸÑÿ¥ŸáŸàÿ± ÿ£Ÿà ÿßŸÑÿ≤ŸÖŸÜ)                             |
| `y`             | ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÑŸä ÿπÿßŸäÿ≤ ÿ™ÿ±ÿ≥ŸÖŸáÿß ŸÉÿÆÿ∑                                 |
| `yerr`          | ŸÖŸÇÿØÿßÿ± ÿßŸÑÿÆÿ∑ÿ£ ÿ£Ÿà ÿßŸÑÿßŸÜÿ≠ÿ±ÿßŸÅ ÿßŸÑŸÖÿπŸäÿßÿ±Ÿä ÿßŸÑŸÑŸä ŸáŸäÿ™ÿ±ÿ≥ŸÖ ŸÉŸÄ Error Bars |
| `fmt` (ÿßÿÆÿ™Ÿäÿßÿ±Ÿä) | ÿ¥ŸÉŸÑ ÿßŸÑÿÆÿ∑ (`-` ÿÆÿ∑ÿå `o` ŸÜŸÇÿßÿ∑ÿå `--` ÿ¥ÿ±ÿ∑ÿ© ŸÖÿ™ŸÇÿ∑ÿπÿ©)              |
| `label`         | ÿßÿ≥ŸÖ ÿßŸÑÿ≥ÿ∑ÿ± ŸÅŸä ÿ≠ÿßŸÑÿ© ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ `legend()`                       |

```python
fig, ax = plt.subplots()

ax.errorbar(df["months"], df["temps"], yerr=df["std_devs"], fmt='-o', color='green', label="Avg Temp")

ax.set_ylabel("Temperature (¬∞C)")
ax.set_xlabel("Month")
ax.legend()
plt.show()
```

## Quantitative comparisons - Scatter Plots

üß© #1. ŸÖÿØÿÆŸÑ ÿ•ŸÑŸâ ÿßŸÑŸÖŸÇÿßÿ±ŸÜÿßÿ™ ÿßŸÑÿ´ŸÜÿßÿ¶Ÿäÿ© (Bi-variate Comparison)

ÿßŸÑŸÄ bar chart ÿ®Ÿäÿ±ŸÉÿ≤ ÿπŸÑŸâ ŸÖÿ™ÿ∫Ÿäÿ± Ÿàÿßÿ≠ÿØÿå ŸÖÿ´ŸÑ ÿπÿØÿØ ÿßŸÑŸÖŸäÿØÿßŸÑŸäÿßÿ™ ŸÑŸÉŸÑ ÿØŸàŸÑÿ©.

ŸÑŸÉŸÜ ŸÑŸà ÿπÿßŸäÿ≤ ÿ™ŸÇÿßÿ±ŸÜ ÿ®ŸäŸÜ ŸÖÿ™ÿ∫Ÿäÿ±ŸäŸÜ ŸÅŸä ŸÜŸÅÿ≥ ÿßŸÑŸàŸÇÿ™ÿü ÿØŸá ÿßÿ≥ŸÖŸá:

‚úÖ "Bi-variate comparison" = ŸÖŸÇÿßÿ±ŸÜÿ© ÿ´ŸÜÿßÿ¶Ÿäÿ© ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ±ÿßÿ™

üü† #2. ŸÖÿß ŸáŸà Scatter Plotÿü

üéØ ÿßŸÑŸáÿØŸÅ:

ŸÜŸÖÿ´ŸëŸÑ ŸÉŸÑ ŸÜŸÇÿ∑ÿ© ŸÖŸÜ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ≠ÿ≥ÿ® ŸÇŸäŸÖÿ™ŸäŸÜ:

ÿßŸÑŸÖÿ≠Ÿàÿ± X ‚Üí ŸäŸÖÿ´ŸÑ ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ± ÿßŸÑÿ£ŸàŸÑ (ŸÖÿ´ŸÑ ÿ™ÿ±ŸÉŸäÿ≤ ÿ´ÿßŸÜŸä ÿ£ŸÉÿ≥ŸäÿØ ÿßŸÑŸÉÿ±ÿ®ŸàŸÜ)

ÿßŸÑŸÖÿ≠Ÿàÿ± Y ‚Üí ŸäŸÖÿ´ŸÑ ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ± ÿßŸÑÿ´ÿßŸÜŸä (ŸÖÿ´ŸÑ ÿØÿ±ÿ¨ÿ© ÿßŸÑÿ≠ÿ±ÿßÿ±ÿ©)

```python
import pandas as pd
import numpy as np

# ÿπÿØÿØ ÿßŸÑÿ¥ŸáŸàÿ± ÿßŸÑŸÑŸä ŸáŸÜÿ¥ÿ™ÿ∫ŸÑ ÿπŸÑŸäŸáÿß
n_months = 50

# ÿ•ŸÜÿ¥ÿßÿ° ÿ™Ÿàÿßÿ±ŸäÿÆ ÿ¥Ÿáÿ±Ÿäÿ© ÿ™ÿ®ÿØÿ£ ŸÖŸÜ ŸäŸÜÿßŸäÿ± 2000
dates = pd.date_range(start="2000-01-01", periods=n_months, freq="MS")

# ÿ™ŸàŸÑŸäÿØ ÿ®ŸäÿßŸÜÿßÿ™ CO2 (ÿ®ÿØÿßŸäÿ© ŸÖŸÜ 370 ÿ®ÿ≤ŸäÿßÿØÿ© ÿ™ÿØÿ±Ÿäÿ¨Ÿäÿ© + ÿ∂Ÿàÿ∂ÿßÿ° ÿπÿ¥Ÿàÿßÿ¶Ÿäÿ© ÿ®ÿ≥Ÿäÿ∑ÿ©)
co2 = 370 + np.linspace(0, 10, n_months) + np.random.normal(0, 0.5, n_months)

# ÿ™ŸàŸÑŸäÿØ ÿ®ŸäÿßŸÜÿßÿ™ ÿØÿ±ÿ¨ÿ© ÿßŸÑÿ≠ÿ±ÿßÿ±ÿ© ÿßŸÑŸÜÿ≥ÿ®Ÿäÿ© (ÿ™ÿ™ÿ∞ÿ®ÿ∞ÿ® ÿπŸÑŸâ ÿ¥ŸÉŸÑ ŸÖŸàÿ¨ÿ© ÿ≥ŸÜŸàŸäÿ©)
relative_temp = 0.1 * np.sin(np.linspace(0, 4 * np.pi, n_months)) + np.random.normal(0, 0.02, n_months)

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿØÿßÿ™ÿß ŸÅÿ±ŸäŸÖ
data = {
    "date": dates,
    "co2": co2,
    "relative_temp": relative_temp
}

df = pd.DataFrame(data)
df.set_index("date", inplace=True)
df
```

```python
fig, ax = plt.subplots()
ax.scatter(df["co2"], df["relative_temp"])
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperature (Celsius)")
plt.show()
```

üîµ #3. ÿ™ÿÆÿµŸäÿµ Scatter Plot + ŸÖŸÇÿßÿ±ŸÜÿ© ÿπŸÇÿØŸäŸÜ ÿ≤ŸÖŸÜŸäŸëŸäŸÜ

üß† ŸÜÿ∂ŸäŸÅ ŸÑŸàŸÜ ŸàÿπŸÜŸàÿßŸÜ ŸÑŸÉŸÑ ŸÖÿ¨ŸÖŸàÿπÿ©:

```python
The_first_two_years = df["2000":"2001"]
The_second_two_years = df["2002":"2004"]

fig, ax = plt.subplots()
ax.scatter(The_first_two_years["co2"], The_first_two_years["relative_temp"], color="red", label="The_first_two_years")
ax.scatter(The_second_two_years["co2"], The_second_two_years["relative_temp"], color="blue", label="The_second_two_years")

ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperature (Celsius)")
ax.legend()
plt.show()
```

üåà #4. ÿ™ÿ±ŸÖŸäÿ≤ ŸÖÿ™ÿ∫ŸäŸëÿ± ÿ´ÿßŸÑÿ´ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿßŸÑŸÑŸàŸÜ (Color as Encoding)

üéØ ÿßŸÑŸáÿØŸÅ:

ŸÜÿ∂ŸäŸÅ ÿ®ÿπÿØ ÿ´ÿßŸÑÿ´ ŸÑŸÑŸÖŸÇÿßÿ±ŸÜÿ© ÿ®ÿØŸàŸÜ ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ŸÖÿ≠Ÿàÿ± ÿ•ÿ∂ÿßŸÅŸä ‚Üí ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿßŸÑŸÑŸàŸÜ!

ŸÖÿ´ŸÑÿßŸã: ŸÜÿÆŸÑŸä ÿßŸÑÿ≤ŸÖŸÜ ŸáŸà ÿßŸÑŸÑŸä ÿ®Ÿäÿ™ÿ≠ŸÉŸÖ ŸÅŸä ŸÑŸàŸÜ ŸÉŸÑ ŸÜŸÇÿ∑ÿ©.

```python
fig, ax = plt.subplots()
scatter = ax.scatter(df["co2"], df["relative_temp"], c=df.index.year, cmap="viridis")

ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperature (Celsius)")
plt.colorbar(scatter, label="Year")
plt.show()
```

c=df.index.year ‚Üí ŸÜÿ≥ÿ™ÿÆÿØŸÖ ÿßŸÑÿ≥ŸÜÿ© ŸÉŸÖÿ™ÿ∫Ÿäÿ± ŸÑŸàŸÜŸä.

cmap="viridis" ‚Üí ÿÆÿ±Ÿäÿ∑ÿ© ÿ£ŸÑŸàÿßŸÜ ŸÖÿ™ÿØÿ±ÿ¨ÿ© ŸÖŸÜ ÿßŸÑÿ£ÿ≤ÿ±ŸÇ ŸÑŸÑÿ£ÿµŸÅÿ±.

plt.colorbar() ‚Üí ÿ™ÿ∂ŸäŸÅ ÿ¥ÿ±Ÿäÿ∑ ÿßŸÑŸÑŸàŸÜ ŸäŸàÿ∂ÿ≠ ÿßŸÑŸÇŸäŸÖ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©.

| ÿßŸÑÿ™ŸÉŸÜŸäŸÉ       | ÿßŸÑŸàÿ∏ŸäŸÅÿ©                                   |
| ------------- | ----------------------------------------- |
| `scatter()`   | ŸÖŸÇÿßÿ±ŸÜÿ© ÿ®ŸäŸÜ ŸÖÿ™ÿ∫Ÿäÿ±ŸäŸÜ (X Ÿà Y)                |

| `legend()`    | ÿ™ÿ≠ÿØŸäÿØ ÿßŸÑÿ£ŸÑŸàÿßŸÜ ŸÅŸä ÿ≠ÿßŸÑÿ© ÿßŸÑŸÖŸÇÿßÿ±ŸÜÿßÿ™ ÿßŸÑŸäÿØŸàŸäÿ©   |

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ÿ•ÿπÿßÿØÿ© ÿ•ŸÜÿ¥ÿßÿ° ÿ®ŸäÿßŸÜÿßÿ™ ŸÖŸäÿØÿßŸÑŸäÿßÿ™ ÿ£ŸàŸÑŸÖÿ®ŸäÿßÿØ 2016
data = {
    "Country": ["United States", "Great Britain", "China", "Russia", "Germany", "Japan", "Egypt"],
    "Gold": [46, 27, 26, 19, 17, 12, 3],
    "Silver": [37, 23, 18, 17, 10, 8, 2],
    "Bronze": [38, 17, 26, 20, 15, 21, 3]
}

df = pd.DataFrame(data)
df.set_index("Country", inplace=True)
```

```python
fig, ax = plt.subplots()

ax.bar(df.index, df["Gold"])                      #  ÿ±ÿ≥ŸÖ ÿ£ÿπŸÖÿØÿ© ŸÑŸÉŸÑ ÿØŸàŸÑÿ©
ax.set_ylabel("Number of medals")
ax.set_title("Olympic Medals - Gold")
ax.set_xticklabels(df.index, rotation=90)       # ÿ™ÿØŸàŸäÿ± ÿ£ÿ≥ŸÖÿßÿ° ÿßŸÑÿØŸàŸÑ ÿπÿ¥ÿßŸÜ ŸÖÿß ÿ™ÿ™ÿØÿßÿÆŸÑÿ¥.
plt.show()
```
