# `Route: Visualization with Matplotlib`

```python
import matplotlib.pyplot as plt
import numpy as np
```

# **Line plot**

```python
x = np.linspace(0, 10, 50)
x
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.title("Sin(x) & values")
plt.xlabel("x values")
plt.ylabel("Sin(x)")
plt.show()
```

```python
# d -> diamod , s-> square , * , o ->circle
plt.plot(x , np.sin(x) , '*')
plt.plot(x , np.cos(x) , 's')
plt.show()
```

```python
# d -> diamod , s-> square , * , o ->circle
plt.plot(x , np.sin(x) , color='blue')
plt.plot(x , np.cos(x) , color='orange')
plt.show()
```

```python
plt.plot(x , color = 'red' , label = 'x values')
plt.plot(x**2 ,color = 'green' , label = 'quadratic')
plt.plot(x**3 , color = 'blue' , label = 'qubic')
plt.title('X vs. X**2 vs. X**3')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

# **Bar plot**

```python
x = np.array(['Class 1' , 'Class 2' , 'Class 3' , 'Class 4' , 'Class 5'])
y = np.array([25 , 40 , 30 , 20 , 35])
```

```python
plt.bar(x, y, color='skyblue')
plt.xlabel('Classes')
plt.ylabel('Num of students');
# plt.show()
```

```python
#horizental bar plot
plt.barh(x , y , color = 'skyblue')
plt.ylabel('Classes')
plt.xlabel('Num of students')
plt.show()
```

```python
data_1 = [20 , 30 , 25 , 28]
data_2 = [12 , 16  ,21 , 13]
x = [0  ,1 , 2 , 3]
plt.bar(x , data_1 , color  ='red')
plt.bar(x , data_2 , color = 'blue')
plt.show()
```

# **Pie chart**

```python
a = np.array([40 , 30 , 30])
b=  np.array(['AI' , 'UI' , 'Web'])
```

```python
plt.pie(a)
plt.show()
```

```python
plt.pie(a , labels = b)
plt.show()
```

```python
a = np.array([40 , 30 , 30])
b=  np.array(['AI' , 'UI' , 'Web'])
color_list = ['red' , 'skyblue' , 'green']
plt.pie(a , labels = b , colors = color_list , autopct = '%1.1f%%')
plt.show()
```

# **Small multiples**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x= np.linspace(0 , 10 , 100)
```

```python
fig, ax = plt.subplots(2, 2)

ax[0,0].plot(x)
ax[0,1].plot(np.sin(x))
ax[1,0].plot(np.cos(x))
ax[1,1].plot(np.tan(x))
plt.tight_layout()
plt.show()
```

```python
data = pd.DataFrame({
    'Month' : ['Jan'  , 'Feb' , 'March' , 'April'],
    'ProductA_Sales' : [120 , 200  , 430 , 280],
    'ProductB_Sales' : [360 , 500 , 230 , 190],
    'ProductC_Sales' : [450 , 234 , 674 ,289],

})
```

```python
plt.figure(figsize = (12 , 5))

fig, ax = plt.subplots(1, 3)
ax[0].plot(data['Month'] , data['ProductA_Sales'] , color = 'red' , marker = 'o')
ax[1].plot(data['Month'] , data['ProductB_Sales'] , color = 'blue' , marker = 'o')
ax[2].plot(data['Month'] , data['ProductC_Sales'] , color = 'green' , marker = 'o')


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (12 , 5))
plt.subplot(1 , 3 , 1 )
plt.plot(data['Month'] , data['ProductA_Sales'] , color = 'red' , marker = 'o' , label = 'Product A')
plt.legend()
plt.title('product A')

plt.subplot(1 , 3 , 2)
plt.plot(data['Month'] , data['ProductB_Sales'] , color = 'blue' , marker = 'o' , label = 'Product B')
plt.legend()
plt.title('product B')

plt.subplot(1 , 3 , 3 )
plt.plot(data['Month'] , data['ProductC_Sales'] , color = 'green' , marker = 'o' , label = 'Product C')
plt.legend()
plt.title('product C')
```

```python
plt.plot(data['Month'] , data['ProductA_Sales'] , color = 'red' , marker = 'o' , label = 'Product A')
plt.plot(data['Month'] , data['ProductB_Sales'] , color = 'blue' , marker = 'o' , label = 'Product B')
plt.plot(data['Month'] , data['ProductC_Sales'] , color = 'green' , marker = 'o' , label = 'Product C')
plt.legend()
plt.title('products')
plt.show()
```

# **Histogram**

```python
# iris dataset
data = pd.read_csv('../Iris.csv')
```

```python
data.head()
```

```python
data.drop(columns = ['Id'] , inplace = True)
```

```python
data.head()
```

```python
plt.hist(data['SepalLengthCm'], bins=15)
plt.title('Histogram of SepalLengthCm')
plt.show()
```

```python
plt.hist(data['SepalLengthCm'], bins=15, edgecolor='black')
plt.title('Histogram of SepalLengthCm')
plt.show()
```
