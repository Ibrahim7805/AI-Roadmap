```python
import plotly.express as px
```

```python
data = px.data.gapminder()
data
```

```python
data = data[data["country"] != "Israel"]
```

```python
"Israel" in data["country"]
```

```python
data.info()
```

# **Scatter plot**

```python
px.scatter(data, y = 'lifeExp', x = 'gdpPercap')
```

```python
px.scatter(data ,  y = 'lifeExp' , x='gdpPercap' , size='pop' , size_max
           =60 , color='continent')
```

```python
px.scatter(data ,  y = 'lifeExp' , x='gdpPercap' , size='pop' , size_max=60 , color='continent' , hover_name='country')
```

```python
px.scatter(data ,  y = 'lifeExp' , x='gdpPercap' , size='pop' , size_max=60 , color='country')
```

```python
px.scatter(data[data['year'] == 2007] ,  y = 'lifeExp' , x='gdpPercap' , size='pop' , size_max=60 , color='country')
```

```python
px.scatter(data, y = 'lifeExp', x = 'gdpPercap', size='pop', size_max=60, color='continent' ,animation_frame='year', hover_name='country' )
```

# **Histogram**

```python
px.histogram(data , 'lifeExp')
```

# **Box Plot**

```python
px.box(data , 'lifeExp')
```

# **Choropleth plot**

```python
px.choropleth(data , locations='iso_alpha')
```

```python
px.choropleth(data , locations='iso_alpha' , color='lifeExp')
```

```python
px.choropleth(data , locations='iso_alpha' , color='lifeExp', scope = 'africa')
```

```python
px.choropleth(data , locations='iso_alpha' , color='lifeExp', scope = 'africa' , hover_name='country')
```

# **Line Chart**

```python
data2 = px.data.stocks()
```

```python
data2
```

```python
px.line(data2 , x='date' , y = 'GOOG')
```

```python
px.line(data2 , x='date' , y = ['GOOG', 'AAPL', 'AMZN', 'FB', 'NFLX', 'MSFT'] )
```

```python
px.line(data2 , x='date' , y = ['GOOG', 'AAPL', 'AMZN', 'FB', 'NFLX', 'MSFT']  , markers='date')
```

# **Area plot**

```python
px.area(data2 , x='date' , y = ['GOOG', 'AAPL', 'AMZN', 'FB', 'NFLX', 'MSFT'] )
```
