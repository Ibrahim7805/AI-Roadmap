‚úÖ 5. ÿ£ŸàŸÑ ŸÖÿ´ÿßŸÑ: Scatter Plot üìç

## **Using pandas with seaborn**

ŸÖÿπÿ∏ŸÖ ÿßŸÑŸÄ Data Scientists ÿ®Ÿäÿ≥ÿ™ÿÆÿØŸÖŸàÿß ŸÖŸÉÿ™ÿ®ÿ© pandas ŸÑÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ÿå ŸàÿØŸá ŸÑÿ£ŸÜ:

```python
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.show()

#  ŸÉŸÑ ŸÜŸÇÿ∑ÿ© ŸÖŸÑŸàŸëŸÜÿ© ÿ≠ÿ≥ÿ® ÿßŸÑŸÇŸäŸÖÿ© ŸÅŸä ÿπŸÖŸàÿØ smoker (Yes / No)
#  Seaborn ÿ®Ÿäÿ∂ŸäŸÅ legend ÿ™ŸÑŸÇÿßÿ¶ŸäŸãÿß
```

"ŸÖÿß ÿßŸÑÿπŸÑÿßŸÇÿ© ÿ®ŸäŸÜ ŸÖÿ™ÿ∫Ÿäÿ±ŸäŸÜ ÿ±ŸÇŸÖŸäŸäŸÜÿü"

ÿßŸÑŸÜŸàÿπ ÿØŸá ŸÖŸÜ ÿßŸÑÿ±ÿ≥ŸàŸÖ ÿßÿ≥ŸÖŸá ŸÅŸä Seaborn: relational plots

ÿ£ÿ¥Ÿáÿ± ŸÖÿ´ÿßŸÑ ÿπŸÑŸäŸáÿß: scatter plots

| ŸáŸÑ ŸáŸÜÿßŸÉ ÿπŸÑÿßŸÇÿ© ÿ®ŸäŸÜ GDP ŸàŸÖÿπÿØŸÑ ÿßŸÑŸÇÿ±ÿßÿ°ÿ©ÿü | GDP vs. literacy   |

ÿ≤Ÿä: "ŸáŸÑ ÿßŸÑÿπŸÑÿßŸÇÿ© ÿ®ŸäŸÜ ÿßŸÑŸÅÿßÿ™Ÿàÿ±ÿ© ŸàÿßŸÑÿ®ŸÇÿ¥Ÿäÿ¥ ÿ™ÿÆÿ™ŸÑŸÅ ÿπŸÜÿØ ÿßŸÑŸÖÿØÿÆŸÜŸäŸÜÿü"

ŸÅŸä ÿßŸÑÿØÿ±ÿ≥ ÿßŸÑŸÑŸä ŸÅÿßÿ™ ÿßÿ≥ÿ™ÿÆÿØŸÖŸÜÿß:

‚úÖ relplot() ŸáŸä:

ÿØÿßŸÑÿ© ÿπÿßŸÖÿ© ŸÑÿπŸÑÿßŸÇÿßÿ™ ÿ±ŸÇŸÖŸäÿ©

ÿ™ŸÇÿØÿ± ÿ™ÿ±ÿ≥ŸÖ:

Scatter plot ‚Üí kind="scatter"

Line plot ‚Üí kind="line" (ŸáŸÜÿ™ÿπŸÑŸÖŸá ÿ®ÿπÿØŸäŸÜ)

üìå ÿßŸÑŸÅÿ±ŸÇ ÿßŸÑŸÉÿ®Ÿäÿ±ÿü

relplot() ÿ™ÿØÿπŸÖ ÿßŸÑÿ±ÿ≥ŸÖ ÿßŸÑŸÖÿ™ÿπÿØÿØ (subplots) ÿ®ÿ≥ŸáŸàŸÑÿ©!

```python
sns.relplot(x="total_bill", y="tip",
            kind="scatter", data=tips)
```

üîÅ 4.5 | ÿßŸÑŸÅÿ±ŸÇ ÿ®ŸäŸÜ scatterplot() Ÿà relplot()
| `scatterplot()`       | `relplot()`                      |
| --------------------- | -------------------------------- |
| ÿ®Ÿäÿ±ÿ≥ŸÖ ÿ±ÿ≥ŸÖ Ÿàÿßÿ≠ÿØ        | ŸäŸÇÿØÿ± Ÿäÿ±ÿ≥ŸÖ ÿπÿØÿ© ÿ±ÿ≥ŸàŸÖ               |
| ŸÖÿÆÿµÿµ ŸÅŸÇÿ∑ ŸÑŸÄ scatter   | ŸÖÿ±ŸÜ: ŸäÿØÿπŸÖ scatter + line         |
| ŸÖÿß ŸäÿØÿπŸÖÿ¥ `col`, `row` | ŸäÿØÿπŸÖ `col`, `row`, `col_wrap`... |


üß± 4.6 | ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ col ŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿ£ÿπŸÖÿØÿ© (Subplots ÿπÿ±ÿ∂ŸäŸãÿß)

```python
sns.relplot(x="total_bill", y="tip",
            col="smoker", kind="scatter",
            data=tips)
```

üß± 4.7 | ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ row ŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿµŸÅŸàŸÅ (Subplots ÿ∑ŸàŸÑŸäŸãÿß)

## **Introduction to line plots**

‚è≥ 6.1 | ŸÖÿß ÿßŸÑŸÅÿ±ŸÇ ÿ®ŸäŸÜ Scatter Plot Ÿà Line Plotÿü

| Scatter Plot                   | Line Plot                              |
| ------------------------------ | -------------------------------------- |
| ŸÉŸÑ ŸÜŸÇÿ∑ÿ© ŸÖÿ≥ÿ™ŸÇŸÑÿ©                 | ÿßŸÑŸÜŸÇÿßÿ∑ ŸÖÿ™ÿ±ÿ™ÿ®ÿ© ÿπÿ®ÿ± ÿßŸÑÿ≤ŸÖŸÜ ÿ£Ÿà ÿ™ÿ±ÿ™Ÿäÿ® ŸÖŸÜÿ∑ŸÇŸä |
| Ÿäÿπÿ®ÿ± ÿπŸÜ ÿπŸÑÿßŸÇÿ© ŸÖÿ®ÿßÿ¥ÿ±ÿ© ÿ®ŸäŸÜ X Ÿà Y | Ÿäÿπÿ®ÿ± ÿπŸÜ ÿ™ÿ∑Ÿàÿ± ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ± ÿ®ŸÖÿ±Ÿàÿ± ÿßŸÑŸàŸÇÿ™       |
| ŸÖÿ´ÿßŸÑ: ÿßŸÑÿ∑ŸàŸÑ ŸÖŸÇÿßÿ®ŸÑ ÿßŸÑŸàÿ≤ŸÜ        | ŸÖÿ´ÿßŸÑ: ÿØÿ±ÿ¨ÿ© ÿ≠ÿ±ÿßÿ±ÿ© ŸÉŸÑ ÿ≥ÿßÿπÿ©               |


```python
import pandas as pd
import numpy as np

# ÿ•ÿπÿØÿßÿØ ÿßŸÑÿπÿ¥Ÿàÿßÿ¶Ÿäÿ© ŸÑŸÑŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑŸÖÿ™ŸÉÿ±ÿ±ÿ©
np.random.seed(42)

# ÿ≥ÿßÿπÿßÿ™ ÿßŸÑŸäŸàŸÖ ŸÖŸÜ 0 ŸÑŸÄ 23
hours = list(range(24))

# ÿßŸÑŸÖŸÜÿßÿ∑ŸÇ ÿßŸÑŸÖÿÆÿ™ŸÑŸÅÿ©
locations = ['North', 'South', 'East', 'West']

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
data = []

for location in locations:
    for hour in hours:
        # ŸÖÿ™Ÿàÿ≥ÿ∑ ŸÖÿ≥ÿ™ŸàŸâ ÿßŸÑÿ™ŸÑŸàÿ´ ÿ®Ÿäÿ≤ŸäÿØ ŸÖÿπ ÿßŸÑŸàŸÇÿ™ + ÿ®ÿπÿ∂ ÿßŸÑÿπÿ¥Ÿàÿßÿ¶Ÿäÿ©
        no2_level = np.random.normal(loc=40 + hour * 0.5, scale=5)
        data.append({
            'hour': hour,
            'location': location,
            'NO_2_mean': round(no2_level, 2)  # ÿ™ŸÇÿ±Ÿäÿ®Ÿá ŸÑŸÖŸÜÿ∑ŸÇŸä
        })

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ DataFrame
df = pd.DataFrame(data)

# ÿπÿ±ÿ∂ ÿ£ŸàŸÑ 5 ÿµŸÅŸàŸÅ
df.head()
```

üîò 6.3 | ÿßŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿÆÿßÿ∑ÿ¶: Scatter Plot

‚ùå ÿ±ÿ≥ŸÖ ÿ¨ŸäÿØÿå ŸÑŸÉŸÜ ŸÖÿ¥ ŸÖÿ´ÿßŸÑŸä ŸÖÿπ ÿ®ŸäÿßŸÜÿßÿ™ ÿ≤ŸÖŸÜŸäÿ©

‚úÖ ÿßŸÑÿ£ŸÅÿ∂ŸÑ ŸÜÿ≥ÿ™ÿÆÿØŸÖ Line Plot.

```python
sns.relplot(x="hour", y="NO_2_mean",
            kind="scatter", data=df)

```

‚úÖ 6.4 | ÿßŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿ£ŸÅÿ∂ŸÑ: Line Plot

ÿ®ŸÜÿ¥ŸàŸÅ ÿßŸÑÿÆÿ∑ ÿ®Ÿäÿ™ÿ≠ÿ±ŸÉ ŸÖÿπ ÿßŸÑŸàŸÇÿ™

ÿ£ÿ≥ŸáŸÑ ŸÅŸä ŸÖÿ™ÿßÿ®ÿπÿ© ÿßŸÑÿßÿ™ÿ¨ÿßŸáÿßÿ™ ŸàÿßŸÑŸÇŸÖŸÖ ŸàÿßŸÑÿßŸÜÿÆŸÅÿßÿ∂ÿßÿ™

```python
sns.relplot(x="hour", y="NO_2_mean",
            kind="line", data=df)
```

üß≠ 6.5 | ÿ™ÿ™ÿ®ÿπ ŸÖÿ¨ŸÖŸàÿπÿßÿ™ ŸÖÿÆÿ™ŸÑŸÅÿ© (Regions)

ŸÑŸà ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ®ÿ™ÿ≠ÿ™ŸàŸä ÿπŸÑŸâ ŸÇÿ±ÿßÿ°ÿßÿ™ ŸÖŸÜ ŸÖŸÜÿßÿ∑ŸÇ ŸÖÿÆÿ™ŸÑŸÅÿ©:

ŸàŸÉŸÑ Ÿàÿßÿ≠ÿØÿ© ÿ®ŸÄ ÿ¥ŸÉŸÑ ÿÆÿ∑ ŸÖÿÆÿ™ŸÑŸÅ

```python
sns.relplot(x="hour", y="NO_2_mean",
            hue="location", style="location",
            kind="line", data=df)
```

## **Count plots and bar plots**

‚úÖ ÿ£ŸàŸÑŸãÿß: ŸäÿπŸÜŸä ÿ•ŸäŸá "Categorical Plot"ÿü

Categorical plot = ÿ±ÿ≥ŸÖ ÿ®ŸäÿßŸÜŸä Ÿäÿπÿ™ŸÖÿØ ÿπŸÑŸâ ŸÖÿ™ÿ∫Ÿäÿ± ÿ™ÿµŸÜŸäŸÅŸä (Categorical Variable)

ŸàÿßŸÑŸÖÿ™ÿ∫Ÿäÿ± ÿßŸÑÿ™ÿµŸÜŸäŸÅŸä = ŸÖÿ™ÿ∫Ÿäÿ± Ÿäÿ≠ÿ™ŸàŸä ÿπŸÑŸâ ŸÇŸäŸÖ ÿ´ÿßÿ®ÿ™ÿ© ŸàŸÖÿ≠ÿØŸàÿØÿ© ŸÖÿ´ŸÑ:

üü° 1. countplot() vs catplot(kind="count")

‚úÖ ÿßŸÑŸÅÿßÿ±ŸÇ ÿßŸÑŸàÿ≠ŸäÿØ ÿ•ŸÜ catplot() Ÿäÿ≥ŸÖÿ≠ ÿ®ŸÖÿ±ŸàŸÜÿ© ÿ£ŸÉÿ´ÿ±:

## **Creating a box plot**

üß† ÿßŸÑŸÅŸÉÿ±ÿ© ÿßŸÑÿ±ÿ¶Ÿäÿ≥Ÿäÿ©:

Box plot ŸáŸà ŸÜŸàÿπ ŸÖŸÜ ÿßŸÑÿ±ÿ≥ŸàŸÖÿßÿ™ ÿßŸÑÿ•ÿ≠ÿµÿßÿ¶Ÿäÿ© ÿ®Ÿäÿπÿ±ÿ∂ ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸÉŸÖŸäÿ© (Quantitative) ÿπÿ®ÿ± ŸÖÿ¨ŸÖŸàÿπÿßÿ™ ŸÖÿÆÿ™ŸÑŸÅÿ© (Categorical Groups).

üìä ÿ¥ŸÉŸÑ ÿßŸÑÿ®ŸàŸÉÿ≥ ÿ®ŸÑŸàÿ™:

ŸÉŸÑ ÿ®ŸàŸÉÿ≥ ÿ®ŸÑŸàÿ™ ÿ®Ÿäÿ™ŸÉŸàŸëŸÜ ŸÖŸÜ:

| ÿßŸÑÿ¨ÿ≤ÿ°                         | ŸÖÿπŸÜÿßŸá                                             |
| ----------------------------- | ------------------------------------------------- |
| **ÿßŸÑŸÄ Box** (ÿßŸÑÿµŸÜÿØŸàŸÇ)         | ŸäŸÖÿ´ŸÑ ÿßŸÑŸÖŸÜÿ∑ŸÇÿ© ŸÖŸÜ **ÿßŸÑŸÄ 25% ÿ•ŸÑŸâ ÿßŸÑŸÄ 75%** (ÿßŸÑŸÄ IQR) |
| **ÿßŸÑÿÆÿ∑ ÿØÿßÿÆŸÑ ÿßŸÑÿµŸÜÿØŸàŸÇ**         | ŸáŸà **ÿßŸÑŸàÿ≥Ÿäÿ∑ (median)**                            |
| **ÿßŸÑŸÄ Whiskers** (ÿßŸÑÿ¥ŸèÿπŸéÿ®)    | ÿ®ÿ™ŸÖÿ´ŸÑ ÿßŸÖÿ™ÿØÿßÿØ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ (ÿ∫ÿßŸÑÿ®Ÿãÿß ŸÑÿ≠ÿØ 1.5√óIQR)        |
| **ÿßŸÑŸÜŸÇÿßÿ∑ ÿßŸÑÿ∑ÿßŸäÿ±ÿ© (Outliers)** | ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÑŸä ÿ®ÿπŸäÿØÿ© ÿ¨ÿØŸãÿß ÿπŸÜ ÿ®ÿßŸÇŸä ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™            |


```python
import seaborn as sns
import matplotlib.pyplot as plt

# ÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸäÿßŸÜÿßÿ™ tips ÿßŸÑÿ¨ÿßŸáÿ≤ÿ©
tips = sns.load_dataset("tips")

# ÿ•ŸÜÿ¥ÿßÿ° box plot
sns.catplot(data=tips, x="time", y="total_bill", kind="box")
plt.show()
```

ŸÑŸÖÿß ÿ™ŸÉŸàŸÜ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ±ŸÇŸÖŸäÿ© ŸÖÿ≥ÿ™ŸÖÿ±ÿ© (ÿ≤Ÿä "ÿßŸÑŸÖÿ±ÿ™ÿ®" ÿ£Ÿà "ÿπÿØÿØ ÿßŸÑŸÅŸàÿßÿ™Ÿäÿ±").

## **Point plots**

‚úÖ ÿ™ÿπÿ±ŸäŸÅ ÿ®ÿ≥Ÿäÿ∑:

ÿ™ÿπÿ±ÿ∂ ŸÖÿ™Ÿàÿ≥ÿ∑ (ÿ£Ÿà ÿ£Ÿä ÿ•ÿ≠ÿµÿßÿ° ÿ¢ÿÆÿ± ŸÖÿ´ŸÑ ÿßŸÑŸàÿ≥Ÿäÿ∑) ŸÑŸÇŸäŸÖÿ© ŸÉŸÖŸäÿ© ŸÑŸÉŸÑ ŸÅÿ¶ÿ© ŸÖŸÜ ŸÅÿ¶ÿ© ÿ™ÿµŸÜŸäŸÅŸäÿ©.

ÿ™ŸÖÿ´ŸÑ ŸÉŸÑ ŸÅÿ¶ÿ© ÿ®ŸÜŸÇÿ∑ÿ©.

ÿ™ÿ∏Ÿáÿ± ŸÅŸàÿßÿµŸÑ ÿßŸÑÿ´ŸÇÿ© (Confidence Intervals) ÿ≠ŸàŸÑ ÿßŸÑŸÜŸÇÿßÿ∑.

üìå ŸÖŸÇÿßÿ±ŸÜÿ© ÿ≥ÿ±Ÿäÿπÿ©:

## **Changing plot style and color**

## **‚úÖ ÿ£ŸàŸÑŸãÿß: ÿ™ÿ∫ŸäŸäÿ± ÿ≥ÿ™ÿßŸäŸÑ ÿßŸÑÿ¥ŸÉŸÑ (Style)**

whitegrid ŸÖŸÜÿßÿ≥ÿ® ÿ¨ÿØŸãÿß ŸÑŸà ÿπÿßŸäÿ≤ ÿ™ÿ¥ŸàŸÅ ÿßŸÑŸÇŸäŸÖ ÿ®ÿØŸÇÿ© ÿ®Ÿàÿ¨ŸàÿØ ÿÆÿ∑Ÿàÿ∑ ÿßŸÑÿ¥ÿ®ŸÉÿ©.

### **‚úÖ ÿ´ÿßŸÜŸäŸãÿß: ÿ™ÿ∫ŸäŸäÿ± ÿßŸÑÿ£ŸÑŸàÿßŸÜ (Color Palette)**

```python
# ÿßÿÆÿ™Ÿäÿßÿ± ŸÖŸÜ ÿßŸÑŸÄ palettes ÿßŸÑÿ¨ÿßŸáÿ≤ÿ©
sns.set_palette("Purples")

# ÿ£Ÿà ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ Ÿàÿßÿ≠ÿØÿ© ŸÖÿ™ÿØÿ±ÿ¨ÿ© (Sequential)
sns.set_palette("Blues")

# ÿ£Ÿà ŸÖÿ™ÿπÿßŸÉÿ≥ÿ© (Diverging)
sns.set_palette("RdBu")

# ÿ£Ÿà ÿπŸÉÿ≥ ÿßŸÑÿ£ŸÑŸàÿßŸÜ ÿ®ÿ•ÿ∂ÿßŸÅÿ© _r
sns.set_palette("RdBu_r")

# ÿ£Ÿà ÿ™ÿπŸÖŸÑ Ÿàÿßÿ≠ÿØÿ© ÿ®ŸÜŸÅÿ≥ŸÉ
sns.set_palette(["#FF0000", "#00FF00", "#0000FF"])
```

ŸàŸÖŸÖŸÉŸÜ ÿ™ŸÖÿ±ÿ±Ÿáÿß ŸÖÿ®ÿßÿ¥ÿ±ÿ© ŸÅŸä palette= ÿØÿßÿÆŸÑ sns.catplot() ÿ£Ÿà sns.countplot() ŸàŸáÿ™ÿ™ÿ∑ÿ®ŸÇ ŸÅŸàÿ±Ÿãÿß.

### **‚úÖ ÿ´ÿßŸÑÿ´Ÿãÿß: ÿ™ÿ∫ŸäŸäÿ± ÿ≠ÿ¨ŸÖ ŸàŸÖŸÇŸäÿßÿ≥ ÿßŸÑÿπŸÜÿßÿµÿ± (Context)**

```python
# paper (ÿµÿ∫Ÿäÿ±)ÿå notebook (ÿßŸÅÿ™ÿ±ÿßÿ∂Ÿä)ÿå talkÿå poster
sns.set_context("talk")
```

### ‚úÖ ŸÖÿ´ÿßŸÑ ŸÉÿßŸÖŸÑ:

```python
import pandas as pd

survey_data = pd.DataFrame({
    "Parents Advice": [
        "Never", "Rarely", "Sometimes", "Often", "Always",
        "Never", "Sometimes", "Always", "Often", "Rarely",
        "Always", "Often", "Never", "Sometimes", "Rarely",
        "Always", "Never", "Often", "Sometimes", "Rarely",
        "Sometimes", "Sometimes", "Always", "Rarely", "Never"
    ]
})
```

## **Adding titles and labels: Part 1**

### **‚úÖ ÿßŸÑÿ∑ÿ±ŸäŸÇÿ© ÿßŸÑÿµÿ≠Ÿäÿ≠ÿ© ŸÑÿ•ÿ∂ÿßŸÅÿ© ÿπŸÜŸàÿßŸÜ ÿπŸÜÿØ ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ catplot**

‚úÖ ŸÑŸäŸá ÿßÿ≥ÿ™ÿÆÿØŸÖŸÜÿß g.fig.suptitle() ŸÖÿ¥ plt.title()ÿü

ŸÑÿ£ŸÜ catplot() ÿ®Ÿäÿ±ÿ¨ÿπ FacetGridÿå ŸàŸáŸà ÿ®Ÿäÿ≠ÿ™ŸàŸä ÿπŸÑŸâ ÿßŸÑÿ¥ŸÉŸÑ ŸÉŸÄ fig.

ÿßŸÑÿπŸÜŸàÿßŸÜ ÿ®ŸäŸÉŸàŸÜ ŸÑŸÑŸÄ figure ŸÉŸÑŸáÿßÿå ŸÖÿ¥ subplot Ÿàÿßÿ≠ÿØÿ©ÿå ŸÅŸÄ ŸÜÿ≥ÿ™ÿÆÿØŸÖ suptitle.

```python
# 1. ÿ£ÿ±ÿ≥ŸÖ ÿßŸÑÿ®ŸÑŸàŸÉ ŸàÿÆÿ≤ŸëŸÜŸá ŸÅŸä ŸÖÿ™ÿ∫Ÿäÿ± g
g = sns.catplot(
    x="Parents Advice",
    data=survey_data,
    kind="count",
    order=["Never", "Rarely", "Sometimes", "Often", "Always"],
    height=4,
    aspect=1.5
)

# 2. ÿ£ÿ∂ŸÅ ÿßŸÑÿπŸÜŸàÿßŸÜ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ suptitle
g.fig.suptitle("Frequency of Parents Giving Advice", y=1.05)

# 3. ÿπÿ±ÿ∂ ÿßŸÑÿ±ÿ≥ŸÖ
plt.show()
```

```python
# 1. ÿ£ÿ±ÿ≥ŸÖ ÿßŸÑÿ®ŸÑŸàŸÉ ŸàÿÆÿ≤ŸëŸÜŸá ŸÅŸä ŸÖÿ™ÿ∫Ÿäÿ± g
g = sns.catplot(
    x="Parents Advice",
    data=survey_data,
    kind="count",
    order=["Never", "Rarely", "Sometimes", "Often", "Always"],
    height=4,
    aspect=1.5
)

# 2. ÿ£ÿ∂ŸÅ ÿßŸÑÿπŸÜŸàÿßŸÜ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ suptitle
g.fig.suptitle("Frequency of Parents Giving Advice", y=1.05)

g.set_axis_labels("Parents' Advice Frequency", "Count")
g.set_xticklabels(rotation=45)
plt.show()
```

| ÿßŸÑŸÖŸáŸÖÿ©                | FacetGrid                           | AxesSubplot                   |
| --------------------- | ----------------------------------- | ----------------------------- |
| ÿßŸÑÿπŸÜŸàÿßŸÜ ÿßŸÑÿ±ÿ¶Ÿäÿ≥Ÿä       | `g.fig.suptitle("Title", y=...)`    | `g.set_title("Title", y=...)` |
| ÿπŸÜŸàÿßŸÜ ŸÅÿ±ÿπŸä (subplots) | `g.set_titles("text {col_name}")`   | ‚ùå ÿ∫Ÿäÿ± ŸÖÿ™ÿßÿ≠ ŸÑÿ£ŸÜŸá ŸÖŸÅÿ±ÿØ          |

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿ¨ÿßŸáÿ≤ÿ©
tips = sns.load_dataset("tips")

# ÿ•ŸÜÿ¥ÿßÿ° catplot ŸÖÿπ ÿ™ŸÇÿ≥ŸäŸÖ ÿ≠ÿ≥ÿ® ÿßŸÑÿ¨ŸÜÿ≥
g = sns.catplot(
    data=tips,
    x="day",              # ŸÖÿ≠Ÿàÿ± ÿ£ŸÅŸÇŸä: ÿ£ŸäÿßŸÖ ÿßŸÑÿ£ÿ≥ÿ®Ÿàÿπ
    y="total_bill",       # ŸÖÿ≠Ÿàÿ± ÿ±ÿ£ÿ≥Ÿä: ÿßŸÑŸÅÿßÿ™Ÿàÿ±ÿ©
    kind="box",           # ŸÜŸàÿπ ÿßŸÑÿ±ÿ≥ŸÖ: box plot
    col="sex",            # ÿπŸÖŸÑ subplot ŸÑŸÉŸÑ ÿ¨ŸÜÿ≥ (ÿ∞ŸÉÿ± - ÿ£ŸÜÿ´Ÿâ)
    height=4, aspect=1    # ÿ≠ÿ¨ŸÖ ŸÉŸÑ subplot
)

# ÿßŸÑÿπŸÜŸàÿßŸÜ ÿßŸÑÿ±ÿ¶Ÿäÿ≥Ÿä
g.fig.suptitle("Distribution of Total Bill by Day and Gender", y=1.05)

# ÿ™ÿ≥ŸÖŸäÿßÿ™ ÿßŸÑŸÖÿ≠ÿßŸàÿ±
g.set(xlabel="Day of the Week", ylabel="Total Bill ($)")

# ÿßŸÑÿπŸÜÿßŸàŸäŸÜ ÿßŸÑŸÅÿ±ÿπŸäÿ© ŸÑŸÉŸÑ subplot
g.set_titles("Gender: {col_name}")

# ÿ™ÿØŸàŸäÿ± ÿßŸÑÿ™ŸÉÿ™ÿßÿ™ ÿπŸÑŸâ ŸÖÿ≠Ÿàÿ± x
plt.xticks(rotation=45)

# ÿπÿ±ÿ∂ ÿßŸÑÿ¥ŸÉŸÑ
plt.show()
```
