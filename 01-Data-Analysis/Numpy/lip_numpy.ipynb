## **Basic Indexing and Slicing**

```python
import numpy as np

# Slicing => [start(0) : end : steps(1) ] not including end

a = np.array(["A", "B", "C", "D", "E", "F"])

print(a.ndim)             # 1
print(a[1])               # 'B'
print(a[1:4])             # ["B", "C", "D"]
print(a[2:])              # ["C", "D", "E", "F"]
print(a[ : : 2])          # ["A", "C", "E"]
print(a[:6])              # ["A", "B", "C", "D", "E", "F"]
```

```python
b = np.array([["A", "B", "X"],
              ["C", "D", "Y"],
              ["E", "F", "Z"]])

print(b.ndim)                        # 2
print(b[1])                          # ["A", "B", "X"]
print("_" * 20)
print(b[1:])                         # [["C", "D", "Y"] ["E", "F", "Z"]]
print("_" * 20)
print(b[:-1])                        # [['A' 'B' 'X'] ['C' 'D' 'Y']]
print("_" * 20)
print(b[ : : 2])                     # [["A", "B", "X"] ["E", "F", "Z"]]
print("_" * 20)
print(b[:])                          # [["A", "B", "X"] ["C", "D", "Y"] ["E", "F", "Z"]]
```
