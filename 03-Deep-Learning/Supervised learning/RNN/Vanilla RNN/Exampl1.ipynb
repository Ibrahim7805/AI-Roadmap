```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import *
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('tsla.us.txt')
```

```python
data
```

```python
data.info()
```

## split data into 80% train  and 20% test

```python
# split data into 80% train  and 20% test

train_data = data.loc[0:int(len(data) *0.8 ) , ['Open']]
test_data = data.loc[int(len(data) *0.8 ): , ['Open']]
```

```python
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_data)
train_scaled
```

## Split train_data into x, y

```python
time_step = 40

x_train = []
y_train = []

for i in range(time_step , len(train_scaled)): #41
    x_train.append(train_scaled[i-time_step:i , 0]) #[41-40:41] =[1:41]
    y_train.append(train_scaled[i , 0])
```

```python
x_train
```

```python
y_train
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
# (n. of samples , time_step  ,features)
x_train.shape  #(1447, 40) ->(1447, 40 , 1)
```

```python
x_train.shape[0]
```

```python
x_train = np.reshape(x_train , (x_train.shape[0] , x_train.shape[1] , 1))
```

```python
x_train.shape
```

## Build Model

```python
rnn_model = Sequential()
rnn_model.add(SimpleRNN(50 ,activation='tanh' , return_sequences=True   ,input_shape=(x_train.shape[1] , 1)))
rnn_model.add(SimpleRNN(30 ,activation='tanh' , return_sequences=True ))
rnn_model.add(Dropout(0.1))
rnn_model.add(SimpleRNN(20 ,activation='tanh' , return_sequences=True ))
rnn_model.add(Dropout(0.1))
rnn_model.add(Dense(1 ))
```

```python
rnn_model.compile(optimizer='adam' , loss='mse')
```

```python
rnn_model.fit(x_train, y_train , epochs = 10)
```

## Split test_data into x, y

```python
time_step = 40
new_test = data.loc[len(data)-len(test_data)-time_step:, ['Open']].values.reshape(-1 , 1)
```

```python
time_step = 40

x_test = []
y_test = []

for i in range(time_step , len(new_test)): #41
    x_test.append(new_test[i-time_step:i , 0]) #[41-40:41] =[1:41]
    y_test.append(new_test[i , 0])
```

```python
x_test = np.array(x_test)
y_test = np.array(y_test)
```
