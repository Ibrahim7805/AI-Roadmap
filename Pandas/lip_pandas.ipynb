## **.agg()**

aggregate = "Ÿäÿ¨ŸÖÿπ/ŸäŸÑÿÆŸëÿµ" ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™

ŸàŸáŸä ÿØÿßŸÑÿ© ŸÅŸä pandas ÿ®ÿ™ÿÆŸÑŸäŸÉ:

ÿ™ÿ∑ÿ®ŸëŸÇ ÿØŸàÿßŸÑ ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿ© (ÿ≤Ÿä mean, sum, min...)

ÿ£Ÿà ÿ≠ÿ™Ÿâ ÿØŸàÿßŸÑ ŸÖÿÆÿµÿµÿ© (ÿ•ŸÜÿ™ ÿ®ÿ™ŸÉÿ™ÿ®Ÿáÿß ÿ®ŸÜŸÅÿ≥ŸÉ)

ÿπŸÑŸâ ÿπŸÖŸàÿØ Ÿàÿßÿ≠ÿØ ÿ£Ÿà ÿπÿØÿ© ÿ£ÿπŸÖÿØÿ©

‚öôÔ∏è 4. ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ .agg() ŸÑÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ŸÖÿÆÿµÿµÿ©

ÿ®ÿ™ÿÆŸÑŸëŸäŸÉ ÿ™ÿ∑ÿ®ŸÇ ÿπÿØÿ© ÿØŸàÿßŸÑ ŸÖÿ±ÿ© Ÿàÿßÿ≠ÿØÿ©

ÿ®ÿ™ŸÇÿ®ŸÑ ÿØŸàÿßŸÑ ŸÖÿÆÿµÿµÿ© ŸÖŸÉÿ™Ÿàÿ®ÿ© ÿ®ÿ•ŸäÿØŸÉ

ÿ®ÿ™ÿ¥ÿ™ÿ∫ŸÑ ÿπŸÑŸâ ÿ£ŸÉÿ™ÿ± ŸÖŸÜ ÿπŸÖŸàÿØ


```python
print(df["age"].agg("mean"))  #ŸÖÿπ ÿØÿßŸÑÿ© ÿ¨ÿßŸáÿ≤ÿ©
print("_"*50)
print(df["age"].agg(["mean", "median", "std"])) #ŸÖÿπ ÿ£ŸÉÿ´ÿ± ŸÖŸÜ ÿØÿßŸÑÿ©
print("_"*50)
print(df[["age", "sex"]].agg(["mean", "median", "max"]))
```

```python
def pct30(column):
    return column.quantile(0.3)
    # ŸáŸÜÿß ÿ®ŸÜÿ≠ÿ≥ÿ® ÿßŸÑŸÇŸäŸÖÿ© ÿßŸÑŸÑŸä 30% ŸÖŸÜ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ£ŸÇŸÑ ŸÖŸÜŸáÿß.

def pct90(col):
    return col.quantile(0.9)

print(df["age"].agg(pct30))  # ÿπŸÑŸä ÿØÿßŸÑŸá ŸÖÿÆÿµÿµŸá
print("_"*50)
print(df[["age", "fbs"]].agg(pct30))
print("_"*50)
print(df[["age", "fbs"]].agg([pct30, pct90 ]))

```
