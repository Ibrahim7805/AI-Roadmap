| البراميتر            | المعنى / الوظيفة                                         | تأثيره على الـ Overfitting                   |
| -------------------- | -------------------------------------------------------- | -------------------------------------------- |
| **max_depth**        | العمق الأقصى للشجرة                                      | ↑ العمق → ↑ overfitting                      |
| **n_estimators**     | عدد الأشجار (الـ learners)                               | ↑ العدد → ↑ الدقة لكن ممكن ↑ overfitting     |
| **min_child_weight** | الحد الأدنى من العينات داخل leaf                         | ↑ القيمة → ↓ overfitting                     |
| **gamma (γ)**        | أقل gain مسموح به عشان نعمل split                        | ↑ القيمة → ↓ overfitting                     |
| **lambda (λ)**       | عامل الـ regularization على الـ weights                  | ↑ القيمة → ↓ overfitting                     |
| **eta (η)**          | الـ learning rate (النسبة اللي بناخدها من كل شجرة جديدة) | ↓ القيمة → ↓ overfitting بس نحتاج أشجار أكتر |
| **subsample**        | نسبة البيانات اللي بتستخدمها كل شجرة                     | ↓ القيمة → ↓ overfitting                     |
| **colsample_bytree** | نسبة الأعمدة اللي بتستخدمها كل شجرة                      | ↓ القيمة → ↓ overfitting                     |
