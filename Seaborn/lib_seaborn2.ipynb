### #4Ô∏è‚É£ ‚Äì Faceting: ÿßŸÑÿ±ÿ≥ŸÖ ÿßŸÑŸÖÿ™ÿπÿØÿØ ÿ®ŸÜÿßÿ°Ÿã ÿπŸÑŸâ ŸÖÿ™ÿ∫Ÿäÿ±ÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©

### üîπ (2) Setting default style with sns.set()

sns.set() ÿ®ÿ™ÿÆŸÑŸä ÿ≥ÿ™ÿßŸäŸÑ Seaborn ŸáŸà ÿßŸÑÿßŸÅÿ™ÿ±ÿßÿ∂Ÿä ŸÅŸä ŸÉŸÑ ÿßŸÑÿ±ÿ≥ŸàŸÖÿßÿ™ ÿßŸÑŸÑŸä ÿ®ÿπÿØ ŸÉÿØŸá.

ÿßŸÑŸÜÿ™Ÿäÿ¨ÿ©ÿü

ÿ£ŸÑŸàÿßŸÜ ŸÖÿ™ŸÜÿßÿ≥ŸÇÿ© ŸàŸáÿßÿØŸäÿ© (Muted colors).

ÿ®Ÿäÿ™ŸÖ ÿßÿ≥ÿ™ÿ®ÿØÿßŸÑ ÿ¥ŸÉŸÑ ÿßŸÑÿ±ÿ≥ŸàŸÖÿßÿ™ ÿßŸÑÿ™ŸÇŸÑŸäÿØŸä ÿ®ÿ™ÿßÿπ Matplotlib ÿ®ÿ±ÿ≥ŸÖÿ© ÿ£ŸÉÿ´ÿ± ÿßÿ≠ÿ™ÿ±ÿßŸÅŸäÿ©.

### **üîπ (4) ÿ•ÿ≤ÿßŸÑÿ© ÿßŸÑÿ≤Ÿàÿßÿ¶ÿØ ÿ®ŸÄ sns.despine()**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ‚¨ÖÔ∏è ÿ™ŸÅÿπŸäŸÑ ÿØÿπŸÖ ŸÉŸàÿØ ÿßŸÑÿ£ŸÑŸàÿßŸÜ ŸÖŸÜ Matplotlib
sns.set(color_codes=True)

# ‚¨ÖÔ∏è ÿ•ŸÜÿ¥ÿßÿ° ÿ®ŸäÿßŸÜÿßÿ™ ÿ®ÿ≥Ÿäÿ∑ÿ© ŸÑŸÅÿ¶ÿßÿ™ ŸÖÿÆÿ™ŸÑŸÅÿ©
data = sns.load_dataset("titanic")

# ‚¨ÖÔ∏è ÿ™ÿ≠ÿØŸäÿØ ÿ™ÿ±ÿ™Ÿäÿ® ÿßŸÑŸÅÿ¶ÿßÿ™ (Categorical)
category_order = ['male', 'female']

# ‚¨ÖÔ∏è ÿßÿÆÿ™Ÿäÿßÿ± Palette ÿßŸÅÿ™ÿ±ÿßÿ∂Ÿäÿ© (Muted)
sns.set_palette("muted")

# ‚¨ÖÔ∏è ÿπÿ±ÿ∂ ÿßŸÑÿ£ŸÑŸàÿßŸÜ ÿßŸÑŸÖŸàÿ¨ŸàÿØÿ© ŸÅŸä ÿßŸÑÿ®ÿßŸÑŸäÿ™ ÿßŸÑŸÖÿÆÿ™ÿßÿ±ÿ©
sns.palplot(sns.color_palette())  # Ÿäÿπÿ±ÿ∂ ÿßŸÑŸÄ Palette ÿßŸÑÿ≠ÿßŸÑŸäÿ©
plt.title("Current Palette - muted")
plt.show()

# ‚¨ÖÔ∏è ÿ±ÿ≥ŸÖ countplot ŸÖÿπ ÿ™ÿ≠ÿØŸäÿØ Palette ŸÖŸÜ ÿßŸÑŸÜŸàÿπ ÿßŸÑÿØÿßÿ¶ÿ±Ÿä (categorical)
sns.set_palette("bright")
sns.countplot(x="sex", data=data, order=category_order)
plt.title("Count of Passengers by Gender (Bright Palette)")
plt.show()

# ‚¨ÖÔ∏è ÿ±ÿ≥ŸÖ ÿ¢ÿÆÿ± ÿ®ŸÑŸàŸÜ ŸÖÿÆÿµÿµ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ color code
sns.histplot(data=data, x="age", color='g', kde=True)
plt.title("Age Distribution (Green color)")
plt.show()

# ‚¨ÖÔ∏è ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿ®ÿßŸÑŸäÿ™ ŸÖŸÜ ÿßŸÑŸÜŸàÿπ Diverging
div_palette = sns.color_palette("coolwarm", as_cmap=False)
sns.palplot(div_palette)
plt.title("Diverging Palette Example (coolwarm)")
plt.show()
```

ŸäÿπŸÜŸä Seaborn ÿ®ŸäÿÆŸÑŸëŸä ÿßŸÑÿ¥ÿ∫ŸÑ ÿ≥ŸáŸÑ Ÿàÿ≥ÿ±Ÿäÿπÿå ŸÑŸÉŸÜ ŸÑŸà ÿπÿßŸäÿ≤ ÿ™ÿÆÿµŸäÿµ ÿ£ŸÉÿ™ÿ± (ÿ≤Ÿä ÿ™ÿ∫ŸäŸäÿ± ÿßŸÑŸÖÿ≠ÿßŸàÿ±ÿå ÿßŸÑÿπŸÜÿßŸàŸäŸÜÿå ÿ£Ÿà ÿ•ÿ∂ÿßŸÅÿ© ÿÆÿ∑Ÿàÿ∑ ŸÖÿ±ÿ¨ÿπŸäÿ©)ÿå ÿ≥ÿßÿπÿ™Ÿáÿß ÿ™ÿ≥ÿ™ÿÆÿØŸÖ Matplotlib.

ÿßŸÑŸÄ Axes ŸáŸà ÿßŸÑÿπŸÜÿµÿ± ÿßŸÑŸÑŸä Seaborn ÿ®Ÿäÿ±ÿ≥ŸÖ ÿπŸÑŸäŸá. ŸÑŸÖÿß ÿ™ŸÜÿ¥ÿ¶ ÿ±ÿ≥ŸÖ ŸÖŸÜ ÿÆŸÑÿßŸÑ plt.subplots()ÿå ÿ®ÿ™ÿ±ÿ¨ÿπŸÑŸÉ ŸÜÿßŸÅÿ∞ÿ© ÿ±ÿ≥ŸÖ (Figure) ŸàŸÖÿ≠Ÿàÿ± ÿ±ÿ≥ŸÖ (Axes). ÿßŸÑŸÖÿ≠Ÿàÿ± ÿØŸá ŸáŸà ÿßŸÑŸÑŸä ÿ®ÿ™ÿ®ÿπÿ™ ÿπŸÑŸäŸá ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™.


### **3. ÿ™ÿÆÿµŸäÿµÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ© ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ Matplotlib**

ÿ®ŸÖÿ¨ÿ±ÿØ ÿ•ŸÜŸÉ ÿ≠ÿµŸÑÿ™ ÿπŸÑŸâ ÿßŸÑŸÄ Axes objectÿå ÿ™ŸÇÿØÿ± ÿ™ÿÆÿµÿµ ÿ£Ÿä ÿπŸÜÿµÿ±:

ax.set_xlabel("..."): ŸÑÿ™ÿ∫ŸäŸäÿ± ÿπŸÜŸàÿßŸÜ ÿßŸÑŸÖÿ≠Ÿàÿ± X.

ax.set_ylabel("..."): ŸÑÿ™ÿ∫ŸäŸäÿ± ÿπŸÜŸàÿßŸÜ ÿßŸÑŸÖÿ≠Ÿàÿ± Y.

ax.set_title("..."): ŸÑÿ•ÿ∂ÿßŸÅÿ© ÿπŸÜŸàÿßŸÜ ŸÑŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿ®ŸäÿßŸÜŸä.

ax.set_xlim([min, max]): ŸÑÿ™ÿ≠ÿØŸäÿØ ŸÖÿØŸâ ÿßŸÑŸÖÿ≠Ÿàÿ± X.

ax.set_ylim([min, max]): ŸÑÿ™ÿ≠ÿØŸäÿØ ŸÖÿØŸâ ÿßŸÑŸÖÿ≠Ÿàÿ± Y.


#### **4. ÿØŸÖÿ¨ ÿ£ŸÉÿ´ÿ± ŸÖŸÜ ÿ±ÿ≥ŸÖ ŸÅŸä ŸÜŸÅÿ≥ ÿßŸÑÿ¥ŸÉŸÑ**

ŸÑŸÖÿß ÿ™ÿ≠ÿ® ÿ™ÿπŸÖŸÑ ŸÖŸÇÿßÿ±ŸÜÿ©ÿå ŸÖŸÖŸÉŸÜ ÿ™ÿ±ÿ≥ŸÖ ÿ£ŸÉÿ™ÿ± ŸÖŸÜ ÿ±ÿ≥ŸÖ ŸÅŸä ŸÜŸÅÿ≥ ÿßŸÑÿ¥ŸÉŸÑ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿ£ŸÉÿ´ÿ± ŸÖŸÜ ŸÖÿ≠Ÿàÿ± Axes.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ®ÿ∑ÿßÿ±ŸäŸÇ (Penguins dataset)
penguins = sns.load_dataset("penguins").dropna()

# ÿ•ŸÜÿ¥ÿßÿ° ŸÖÿ≠Ÿàÿ±ŸäŸÜ ŸÑŸÑÿ±ÿ≥ŸÖ ŸÅŸä ÿµŸÅ Ÿàÿßÿ≠ÿØ
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

# 1Ô∏è‚É£ ÿßŸÑŸÖÿ≠Ÿàÿ± ÿßŸÑÿ£ŸàŸÑ: ŸÉŸÑ ÿßŸÑÿ£ŸÜŸàÿßÿπ
sns.histplot(data=penguins, x="flipper_length_mm", ax=ax1)
ax1.set_title("Flipper Length - All Penguins", fontsize=14)
ax1.set_xlabel("Flipper Length (mm)")
ax1.set_ylabel("Count")
ax1.axvline(200, color='red', linestyle='--', linewidth=2, label='Limit')
ax1.legend()

# 2Ô∏è‚É£ ÿßŸÑŸÖÿ≠Ÿàÿ± ÿßŸÑÿ´ÿßŸÜŸä: ŸÜŸàÿπ ŸÖÿπŸäŸÜ - Adelie
sns.histplot(data=penguins.query("species == 'Adelie'"), x="flipper_length_mm", ax=ax2, color="orange")
ax2.set_title("Flipper Length - Adelie Penguins", fontsize=14)
ax2.set_xlabel("Flipper Length (mm)")
ax2.axvline(200, color='red', linestyle='--', linewidth=2)

# ÿ™ŸÜÿ≥ŸäŸÇ ÿßŸÑÿ¥ŸÉŸÑ ÿßŸÑŸÜŸáÿßÿ¶Ÿä
plt.suptitle("Comparison of Flipper Lengths", fontsize=16)
plt.tight_layout()
plt.show()

```

## **Categorical Plot Types**

‚úÖ ÿ£ŸàŸÑŸãÿß: ŸÖÿß ŸáŸä ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿµŸÜŸäŸÅŸäÿ©ÿü (Categorical Data)

ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿµŸÜŸäŸÅŸäÿ©: ŸáŸä ÿ®ŸäÿßŸÜÿßÿ™ ÿ∫Ÿäÿ± ÿ±ŸÇŸÖŸäÿ© ÿ™ÿ≠ÿ™ŸàŸä ÿπŸÑŸâ ÿπÿØÿØ ŸÖÿ≠ÿØŸàÿØ ŸÖŸÜ ÿßŸÑŸÇŸäŸÖ (ŸÖÿ´ŸÑ: ŸÜŸàÿπ ÿßŸÑŸÖÿ±ÿ∂ÿå ÿßÿ≥ŸÖ ÿßŸÑÿØŸàŸÑÿ©ÿå ŸÅÿ¶ÿ© ÿßŸÑŸÖŸÜÿ™ÿ¨... ÿ•ŸÑÿÆ).

Seaborn Ÿäÿ™Ÿäÿ≠ ŸÑŸÉ ÿπÿ±ÿ∂ Ÿáÿ∞Ÿá ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ¨ŸÜÿ®Ÿãÿß ÿ•ŸÑŸâ ÿ¨ŸÜÿ® ŸÖÿπ ŸÇŸäŸÖ ÿπÿØÿØŸäÿ© (ŸÖÿ´ŸÑ: ÿπÿØÿØ ÿßŸÑŸÖÿ±ÿ∂Ÿâ ŸÅŸä ŸÉŸÑ ŸÅÿ¶ÿ©ÿå ŸÖÿ™Ÿàÿ≥ÿ∑ ÿßŸÑÿ™ŸÉŸÑŸÅÿ© ŸÑŸÉŸÑ ÿ≠ÿßŸÑÿ©...).

üìä ÿ£ŸÜŸàÿßÿπ ÿßŸÑÿ±ÿ≥ŸàŸÖ ÿßŸÑÿ®ŸäÿßŸÜŸäÿ© ÿßŸÑÿ™ÿµŸÜŸäŸÅŸäÿ©

Seaborn ÿ®ŸäŸÇÿ≥ŸÖ ÿßŸÑÿ±ÿ≥ŸàŸÖ ÿØŸä ÿ•ŸÑŸâ 3 ÿ£ŸÜŸàÿßÿπ:

### **ÿπÿ±ÿ∂ ŸÉŸÑ ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÅÿ±ÿØŸäÿ© (individual observations)**

### **üü¢ ÿ£ŸÖÿ´ŸÑÿ©: stripplot(),  swarmplot()**

### üéØ 1. **stripplot**()



#### **üß† ÿßŸÑŸÅŸÉÿ±ÿ©:**

ÿßŸÑŸÄ stripplot() ÿ®Ÿäÿπÿ±ÿ∂ ŸÉŸÑ ŸÜŸÇÿ∑ÿ© ÿ®ŸäÿßŸÜÿßÿ™ ŸÅŸä ÿ¥ŸÉŸÑ ŸÜŸÇÿßÿ∑ ŸÖÿ®ÿπÿ´ÿ±ÿ© (dots) ÿπŸÑŸâ ŸÖÿ≠Ÿàÿ± ŸÖÿπŸäŸÜÿå ÿ®ŸäŸÉŸàŸÜ ŸÖŸÅŸäÿØ ÿπŸÑÿ¥ÿßŸÜ ÿ™ÿ¥ŸàŸÅ ÿßŸÑÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑŸÅÿπŸÑŸä ŸÑŸÑÿ®ŸäÿßŸÜÿßÿ™ÿå ÿÆÿßÿµÿ© ŸÖÿπ ÿπÿØÿØ ŸÇŸÑŸäŸÑ ŸÜÿ≥ÿ®ŸäŸãÿß ŸÖŸÜ ÿßŸÑŸÇŸäŸÖ.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
# ÿ±ÿ≥ŸÖ stripplot
sns.stripplot(x="total_bill", y="day", data=tips, jitter=True)
# ÿπÿ±ÿ∂ ÿßŸÑÿ±ÿ≥ŸÖ
plt.title("Total Bill Distribution per Day (Stripplot)")
plt.show()

# ÿ®ŸÜÿ≥ÿ™ÿÆÿØŸÖ jitter=True ÿπŸÑÿ¥ÿßŸÜ ŸÜÿπŸÖŸÑ "ÿßŸáÿ™ÿ≤ÿßÿ≤" ÿ®ÿ≥Ÿäÿ∑ ŸÅŸä ÿßÿ™ÿ¨ÿßŸá X ÿ£Ÿà Yÿå ŸÅŸäÿ≥ŸáŸÑ ÿ™ŸÖŸäŸäÿ≤ ÿßŸÑŸÜŸÇÿßÿ∑.
```

### **üéØ 2. swarmplot()**


#### **üß† ÿßŸÑŸÅŸÉÿ±ÿ©:**

ÿßŸÑŸÄ swarmplot() ÿ≤Ÿä stripplot()ÿå ŸÑŸÉŸÜŸá ÿ®ŸäŸÖŸÜÿπ ÿ™ÿ±ÿßŸÉÿ® ÿßŸÑŸÜŸÇÿßÿ∑ ÿ™ŸÑŸÇÿßÿ¶ŸäŸãÿß ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿÆŸàÿßÿ±ÿ≤ŸÖŸäÿ© ÿ™ŸÜÿ≥ŸäŸÇ ÿ∞ŸÉŸäÿ©.



```python
sns.swarmplot(x="total_bill", y="day", data=tips)

plt.title("Total Bill Distribution per Day (Swarmplot)")
plt.show()
```

### **üéØ 3. boxplot()**


#### **üß† ÿßŸÑŸÅŸÉÿ±ÿ©:**

ÿßŸÑŸÖÿ±ÿ®ÿπ: ÿßŸÑÿ±ÿ®ÿπŸäŸÜ ÿßŸÑÿ´ÿßŸÜŸä ŸàÿßŸÑÿ´ÿßŸÑÿ´ (IQR)

## **Matrix Plots**


### **‚úÖ ŸÖÿß ŸáŸä Matrix Plotsÿü**

ŸáŸä ÿ±ÿ≥ŸàŸÖÿßÿ™ ÿ™Ÿèÿπÿ±ÿ∂ ŸÅŸä ÿ¥ŸÉŸÑ ŸÖÿµŸÅŸàŸÅÿ© (Matrix)ÿå ŸÉŸÑ ÿÆŸÑŸäÿ© ŸÅŸäŸáÿß ÿ™ŸÖÿ´ŸÑ ÿπŸÑÿßŸÇÿ© ÿ£Ÿà ŸÇŸäŸÖÿ© ÿ®ŸäŸÜ ÿµŸÅ ŸàÿπŸÖŸàÿØ ŸÖÿπŸäŸÜŸäŸÜ. ÿ£ÿ¥Ÿáÿ± ŸÜŸàÿπ ŸÖŸÜŸáÿß ŸáŸà:

### **üî• Heatmap = ÿÆÿ±Ÿäÿ∑ÿ© ÿ≠ÿ±ÿßÿ±ÿ©**

ÿ™ÿ≥ÿ™ÿÆÿØŸÖ ÿßŸÑÿ£ŸÑŸàÿßŸÜ ŸÑÿ™Ÿàÿ∂Ÿäÿ≠ ÿßŸÑŸÇŸäŸÖ ÿßŸÑÿ±ŸÇŸÖŸäÿ©.

ŸÖŸÅŸäÿØÿ© ÿ¨ÿØŸãÿß ŸÑÿ±ÿ§Ÿäÿ© ÿßŸÑÿ£ŸÜŸÖÿßÿ∑ (patterns) ŸàÿßŸÑÿßÿ™ÿ¨ÿßŸáÿßÿ™ ÿ®ÿ≥ÿ±ÿπÿ© ŸÅŸä ÿßŸÑÿØÿßÿ™ÿß.


### **üßÆ 2. Getting Data in the Right Format**

‚ú≥Ô∏è ŸÇÿ®ŸÑ ŸÖÿß ÿ™ÿπŸÖŸÑ heatmapÿå ŸÑÿßÿ≤ŸÖ ÿ™ÿ¨Ÿáÿ≤ ÿ®ŸäÿßŸÜÿßÿ™ŸÉ ŸÅŸä ÿ¥ŸÉŸÑ matrix (ÿ¨ÿØŸàŸÑ ŸÖŸÜÿ∏ŸÖ)
ŸÅŸä Seabornÿå ÿ∫ÿßŸÑÿ®Ÿãÿß ŸÜÿ≥ÿ™ÿÆÿØŸÖ:

#### pd.crosstab(index, columns, values, aggfunc)



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸäÿØŸàŸäŸãÿß
data = {
    'month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Mar', 'Mar', 'Mar',
              'Jun', 'Jun', 'Jun', 'Sep', 'Sep', 'Sep'],
    'day': ['Monday', 'Wednesday', 'Saturday'] * 5,
    'rentals': [105, 130, 200, 90, 110, 180, 120, 140, 190, 170, 180, 250, 160, 170, 240]
}

df = pd.DataFrame(data)

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ matrix ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ  crosstabÿßŸà ()pivot_table
pivot_table = pd.crosstab(index=df['day'], columns=df['month'], values=df['rentals'], aggfunc='mean').round(0)

# ÿπÿ±ÿ∂ ÿ£ŸàŸÑ 5 ÿµŸÅŸàŸÅ ŸÖŸÜ ÿßŸÑŸÖÿµŸÅŸàŸÅÿ©
pivot_table
```


### **üî• 3. Build a Heatmap**

‚ú≥Ô∏è ÿßÿ≥ÿ™ÿÆÿØŸÖ sns.heatmap() ŸÑÿπÿ±ÿ∂ ÿßŸÑÿ¨ÿØŸàŸÑ ŸÅŸä ÿµŸàÿ±ÿ© ÿ£ŸÑŸàÿßŸÜ

```python
sns.heatmap(pivot_table)
plt.title("Average Rentals by Day and Month")
plt.show()
```


### **üé® 4. Customize a Heatmap**

‚úÖ ÿ™ŸÇÿØÿ± ÿ™ÿÆÿµÿµ ÿßŸÑÿ¥ŸÉŸÑ ÿ®ÿßŸÑÿÆŸäÿßÿ±ÿßÿ™ ÿßŸÑÿ™ÿßŸÑŸäÿ©:

| ÿßŸÑÿÆŸäÿßÿ±           | ŸÖÿπŸÜÿßŸá                         |
| ---------------- | ----------------------------- |
| `annot=True`     | Ÿäÿπÿ±ÿ∂ ÿßŸÑŸÇŸäŸÖ ÿßŸÑÿ±ŸÇŸÖŸäÿ© ŸÅŸä ŸÉŸÑ ÿÆŸÑŸäÿ© |
| `fmt="d"`        | Ÿäÿπÿ±ÿ∂ ÿßŸÑÿ£ÿ±ŸÇÿßŸÖ ŸÉŸÄ ÿ£ÿπÿØÿßÿØ ÿµÿ≠Ÿäÿ≠ÿ©   |
| `cmap="YlGnBu"`  | Ÿäÿ≠ÿØÿØ ÿ¥ŸÉŸÑ ÿßŸÑÿ£ŸÑŸàÿßŸÜ              |
| `cbar=False`     | ŸäÿÆŸÅŸä ÿ¥ÿ±Ÿäÿ∑ ÿßŸÑÿ£ŸÑŸàÿßŸÜ ÿßŸÑÿ¨ÿßŸÜÿ®Ÿä     |
| `linewidths=0.5` | Ÿäÿ∂ŸäŸÅ ŸÅŸàÿßÿµŸÑ ÿ®ŸäŸÜ ÿßŸÑÿÆŸÑÿßŸäÿß        |


```python
sns.heatmap(pivot_table, annot=True, fmt="2f", cmap="YlGnBu"	, linewidths=0.5	)
plt.title("Average Rentals by Day and Month")
plt.show()
```

### üü° 5. **Centering a Heatmap**

‚úÖ ÿ™ŸÇÿØÿ± ÿ™ÿÆŸÑŸä ÿßŸÑÿ£ŸÑŸàÿßŸÜ ÿ™ÿ™ŸÖÿ±ŸÉÿ≤ ÿ≠ŸàÿßŸÑŸäŸÜ ŸÇŸäŸÖÿ© ŸÖÿπŸäŸÜÿ©:

```python
sns.heatmap(pivot_table, center_value = pivot_table.iloc[1, 4]
)  # ŸäŸàŸÖ ÿßŸÑÿ≥ÿ®ÿ™ (6)ÿå ÿ¥Ÿáÿ± ÿ≥ÿ®ÿ™ŸÖÿ®ÿ± (9)
plt.title("Centered Heatmap")
plt.show()
```

```python
corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.DataFrame({
    "tuition": [10000, 12000, 15000, 8000, 9500, 18000, 21000, 17000],
    "degree": ["Bachelor", "Bachelor", "Graduate", "Graduate", "Bachelor", "Graduate", "Bachelor", "Graduate"],
    "region": ["East", "West", "East", "West", "East", "West", "East", "West"],
    "control_type": ["Public", "Private", "Private", "Public", "Private", "Public", "Public", "Private"]
})
```

###  üß© **FacetGrid ‚Äì ÿßŸÑŸÖÿ±ŸÜ ŸàÿßŸÑÿ£ŸÇŸàŸâ**

```python
g = sns.FacetGrid(df, col="degree", row="region", hue="control_type")
g.map(sns.boxplot, "tuition")
```

### **üê± catplot() ‚Äì ÿ£ÿ≥ŸáŸÑ ÿ∑ÿ±ŸäŸÇÿ© ŸÑŸÑŸÖÿÆÿ∑ÿ∑ÿßÿ™**

```python
sns.catplot(data=df, x="degree", y="tuition", col="region", kind="box")
```

### **üìà lmplot() ‚Äì ŸÑŸÑŸÖÿÆÿ∑ÿ∑ÿßÿ™ ÿßŸÑÿßŸÜÿ≠ÿØÿßÿ±Ÿäÿ©**

```python
sns.lmplot(data=df, x="control_type", y="tuition", col="degree", row="region", hue="control", fit_reg=True)
```
