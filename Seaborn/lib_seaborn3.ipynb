# **Bar plot**

```python
import matplotlib.pyplot as plt
plt.bar(data['day'], data['total_bill'])
```

```python
sns.barplot(x= 'day'  , y = 'total_bill', data = data )
```

```python
sns.barplot(x= 'day'  , y = 'total_bill', data = data, estimator = sum )
# default == estimator = mean
```

```python
sns.barplot(x='sex', y = 'total_bill', data = data, estimator = sum)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = data, estimator = sum, hue = 'smoker')
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = data, estimator = sum, hue = 'day')
```

```python
sns.barplot(x= 'day'  , y = 'total_bill', data = data , estimator = sum  , hue = 'time')
```

```python
sns.barplot(x= 'day'  , y = 'tip', data = data , estimator = sum  , hue = 'smoker')
```

# **Count plot**

```python
data['sex'].value_counts()
```

```python
sns.countplot(x = 'sex', data = data)
```

```python
sns.countplot(x = 'sex', data = data, hue = 'time')
```

# **Scatter plot**

```python
sns.scatterplot(x = 'total_bill' , y = 'tip' , data = data)
```

```python
sns.scatterplot(x = 'total_bill' , y = 'tip' , data = data , hue = 'smoker')
```

```python
sns.scatterplot(x = 'total_bill' , y = 'tip' , data = data , hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , data = data , hue = 'sex')
```

# **Box plot**

```python
sns.boxplot(x = 'day' , y = 'total_bill'  , data = data)
```

```python
sns.boxplot(x = 'day' , y = 'total_bill'  , data = data , hue = 'sex')
```

```python
plt.boxplot(data['total_bill']);
```

# **Histogram**

```python
sns.histplot(data['total_bill'])
```

```python
sns.histplot(data['total_bill'], kde= True)
```

Kde plot

```python
sns.kdeplot(data['total_bill'] , fill = True , color = 'red')
```

# **Pair plot**

```python
df = sns.load_dataset('tips')
```

```python
sns.pairplot(data = df)
```

```python
sns.pairplot(data = df, hue = 'day')
```

```python
sns.pairplot(data = df, hue = 'time')
```

# HeatMap

```python
sns.heatmap(data.isnull(), cbar=False)
plt.show()
```

# **kdeplot**

```python
sns.kdeplot(x=data['fare'], fill=True)

# Mean Line
plt.vlines(data['fare'].mean(), 0, 0.02, linestyles= 'dotted', colors='black')

plt.xlim(-50, 200)
plt.show()
```
